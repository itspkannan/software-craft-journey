# Arrays & Strings - Concepts & Use Cases

**Arrays** and **strings** are the most fundamental data structures. Arrays store elements in contiguous memory, while strings are sequences of characters (often implemented as character arrays).

---
## Arrays - Core Concepts

### Memory Layout
```
Index:    0     1     2     3     4
        +-----+-----+-----+-----+-----+
Array:  | 10  | 20  | 30  | 40  | 50  |
        +-----+-----+-----+-----+-----+
Address: 100   104   108   112   116  (4 bytes each)

Address of arr[i] = base_address + (i × element_size)
```

### Key Properties
| Property | Description |
|----------|-------------|
| **Contiguous** | Elements stored in consecutive memory |
| **Fixed Size** | Size determined at creation (in most languages) |
| **Random Access** | O(1) access by index |
| **Homogeneous** | All elements same type (in typed languages) |

---
## Python Lists vs Arrays

```python
# Python list (dynamic array)
lst = [1, 2, 3, 4, 5]

# True array (fixed type)
import array
arr = array.array('i', [1, 2, 3, 4, 5])

# NumPy array (for numerical computing)
import numpy as np
np_arr = np.array([1, 2, 3, 4, 5])
```

| Feature | Python List | array.array | NumPy |
|---------|-------------|-------------|-------|
| Mixed types | Yes | No | No |
| Dynamic size | Yes | Yes | Fixed |
| Memory efficient | No | Yes | Yes |
| Fast math ops | No | No | Yes |

````python
# Python list operations
arr = [1, 2, 3, 4, 5]

# Access - O(1)
print(f"arr[2] = {arr[2]}")

# Append - O(1) amortized
arr.append(6)
print(f"After append: {arr}")

# Insert - O(n)
arr.insert(0, 0)
print(f"After insert at 0: {arr}")

# Delete - O(n)
arr.pop(0)
print(f"After pop(0): {arr}")

# Slicing - O(k) where k is slice size
print(f"arr[1:4] = {arr[1:4]}")
````

## Time & Space Complexity

### Array Operations

| Operation | Time | Notes |
|-----------|------|-------|
| Access by index | O(1) | Direct memory calculation |
| Search (unsorted) | O(n) | Linear scan |
| Search (sorted) | O(log n) | Binary search |
| Insert at end | O(1)* | Amortized for dynamic arrays |
| Insert at index | O(n) | Shift elements |
| Delete at end | O(1) | No shift needed |
| Delete at index | O(n) | Shift elements |

### String Operations (Python)

| Operation | Time | Notes |
|-----------|------|-------|
| Access char | O(1) | Index access |
| Concatenation | O(n+m) | Creates new string |
| Substring | O(k) | k = substring length |
| Find/Search | O(n×m) | n = text, m = pattern |

## Common Patterns

### 1. Two Pointers

Two indices moving through array, often from opposite ends.

```
Array: [1, 2, 3, 4, 5]
        ↑           ↑
       left       right
```

**Use cases**: Palindrome check, two sum (sorted), container with most water

````python
def two_sum_sorted(nums, target):
    """Find two numbers that sum to target (sorted array)."""
    left, right = 0, len(nums) - 1
    
    while left < right:
        total = nums[left] + nums[right]
        if total == target:
            return [left, right]
        elif total < target:
            left += 1
        else:
            right -= 1
    
    return [-1, -1]

print(f"Two sum [1,2,3,4,5] target=7: {two_sum_sorted([1,2,3,4,5], 7)}")
````

### 2. Sliding Window

Window of fixed or variable size sliding through array.

```
Array: [1, 2, 3, 4, 5, 6, 7]
        [-----]              Window size 3
           [-----]           Slide right
              [-----]
```

**Use cases**: Max sum subarray, longest substring without repeating chars

````python
def max_sum_subarray(nums, k):
    """Find max sum of subarray of size k."""
    window_sum = sum(nums[:k])
    max_sum = window_sum
    
    for i in range(k, len(nums)):
        window_sum += nums[i] - nums[i - k]  # Slide window
        max_sum = max(max_sum, window_sum)
    
    return max_sum

print(f"Max sum k=3 in [1,4,2,10,2,3,1,0,20]: {max_sum_subarray([1,4,2,10,2,3,1,0,20], 3)}")
````

### 3. Prefix Sum

Precompute cumulative sums for O(1) range queries.

```
Array:      [1, 2, 3, 4,  5]
Prefix Sum: [1, 3, 6, 10, 15]

Sum(i to j) = prefix[j] - prefix[i-1]
Sum(1 to 3) = prefix[3] - prefix[0] = 10 - 1 = 9  (2+3+4)
```

````python
def build_prefix_sum(nums):
    prefix = [0] * (len(nums) + 1)
    for i, num in enumerate(nums):
        prefix[i + 1] = prefix[i] + num
    return prefix

def range_sum(prefix, i, j):
    return prefix[j + 1] - prefix[i]

nums = [1, 2, 3, 4, 5]
prefix = build_prefix_sum(nums)
print(f"Array: {nums}")
print(f"Prefix: {prefix}")
print(f"Sum(1,3): {range_sum(prefix, 1, 3)}")
````

### 4. Hash Map for Lookup

Use dictionary for O(1) lookup of seen elements.

**Use cases**: Two sum, subarray sum equals k, anagram grouping

````python
def two_sum(nums, target):
    """Find two numbers that sum to target."""
    seen = {}  # value -> index
    
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    
    return [-1, -1]

print(f"Two sum [2,7,11,15] target=9: {two_sum([2,7,11,15], 9)}")
````

## String-Specific Techniques

### Immutability in Python
```python
s = "hello"
s[0] = 'H'  # ERROR! Strings are immutable

# Instead, create new string
s = 'H' + s[1:]  # "Hello"

# Or use list for modifications
chars = list(s)
chars[0] = 'H'
s = ''.join(chars)
```

### Character Frequency
```python
from collections import Counter
freq = Counter("aabbcc")  # {'a': 2, 'b': 2, 'c': 2}
```

### ASCII Values
```python
ord('a')  # 97
chr(97)   # 'a'
ord('A')  # 65
```

````python
# String techniques demo
from collections import Counter

# Character frequency
s = "programming"
freq = Counter(s)
print(f"Frequency of '{s}': {dict(freq)}")

# Check anagram
def is_anagram(s1, s2):
    return Counter(s1) == Counter(s2)

print(f"'listen' anagram of 'silent': {is_anagram('listen', 'silent')}")

# Reverse string
s = "hello"
print(f"Reverse of '{s}': {s[::-1]}")
````

## Key Tips

1. **Clarify constraints**: Size limits, character set, sorted?
2. **Consider edge cases**: Empty array, single element, duplicates
3. **Two pointers** for sorted arrays or palindrome problems
4. **Sliding window** for subarray/substring with constraint
5. **Hash map** to trade space for time (O(n) → O(1) lookup)
6. **Sort first** if it helps (O(n log n) preprocessing)
7. **Strings are immutable** in Python - use list for modifications
8. **In-place** means O(1) extra space
