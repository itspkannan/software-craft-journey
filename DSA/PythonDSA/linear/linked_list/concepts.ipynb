# Linked Lists - Concepts & Use Cases

A **linked list** is a linear data structure where elements are stored in nodes, each containing data and a reference (pointer) to the next node.

---
## Core Concepts

### Singly Linked List
```
head
  ↓
+------+    +------+    +------+    +------+
|  1   | -> |  2   | -> |  3   | -> |  4   | -> None
+------+    +------+    +------+    +------+
```

### Doubly Linked List
```
head                                          tail
  ↓                                             ↓
+------+    +------+    +------+    +------+
|  1   | ⇄ |  2   | ⇄ |  3   | ⇄ |  4   |
+------+    +------+    +------+    +------+
```

### Circular Linked List
```
    +------+    +------+    +------+
┌-> |  1   | -> |  2   | -> |  3   | -┐
|   +------+    +------+    +------+  |
└─────────────────────────────────────┘
```

````python
from typing import Optional

# Singly Linked List Node
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def __repr__(self):
        vals = []
        node = self
        while node:
            vals.append(str(node.val))
            node = node.next
        return ' -> '.join(vals)

# Doubly Linked List Node
class DListNode:
    def __init__(self, val=0, prev=None, next=None):
        self.val = val
        self.prev = prev
        self.next = next

# Helper to build list
def build_list(vals):
    if not vals:
        return None
    head = ListNode(vals[0])
    curr = head
    for val in vals[1:]:
        curr.next = ListNode(val)
        curr = curr.next
    return head

# Demo
head = build_list([1, 2, 3, 4, 5])
print(f"List: {head}")
````

## Array vs Linked List

| Operation | Array | Linked List |
|-----------|-------|-------------|
| Access by index | O(1) | O(n) |
| Search | O(n) | O(n) |
| Insert at head | O(n) | O(1) |
| Insert at tail | O(1)* | O(n) or O(1)** |
| Insert at middle | O(n) | O(1)*** |
| Delete at head | O(n) | O(1) |
| Delete at tail | O(1) | O(n) or O(1)** |
| Memory | Contiguous | Scattered |
| Cache locality | Good | Poor |

*Amortized for dynamic arrays  
**With tail pointer  
***If you have reference to the node

## Common Operations

````python
# Insert at head - O(1)
def insert_head(head, val):
    new_node = ListNode(val)
    new_node.next = head
    return new_node

# Insert at tail - O(n)
def insert_tail(head, val):
    if not head:
        return ListNode(val)
    curr = head
    while curr.next:
        curr = curr.next
    curr.next = ListNode(val)
    return head

# Delete node with value - O(n)
def delete_value(head, val):
    dummy = ListNode(0, head)
    curr = dummy
    while curr.next:
        if curr.next.val == val:
            curr.next = curr.next.next
            break
        curr = curr.next
    return dummy.next

# Demo
head = build_list([1, 2, 3])
print(f"Original: {head}")
head = insert_head(head, 0)
print(f"Insert 0 at head: {head}")
head = insert_tail(head, 4)
print(f"Insert 4 at tail: {head}")
head = delete_value(head, 2)
print(f"Delete 2: {head}")
````

## Common Patterns

### 1. Dummy Head

Simplifies edge cases (empty list, delete head).

```python
dummy = ListNode(0, head)
# ... operations ...
return dummy.next  # New head
```

### 2. Fast and Slow Pointers

Two pointers moving at different speeds.

```
slow: 1 step at a time
fast: 2 steps at a time

1 -> 2 -> 3 -> 4 -> 5
s         f           (after 2 moves, slow at middle)
```

**Use cases**: Find middle, detect cycle, find cycle start

````python
def find_middle(head):
    """Find middle node using fast/slow pointers."""
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow

def has_cycle(head):
    """Detect cycle using Floyd's algorithm."""
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

# Demo
head = build_list([1, 2, 3, 4, 5])
print(f"List: {head}")
print(f"Middle: {find_middle(head).val}")
print(f"Has cycle: {has_cycle(head)}")
````

### 3. Reverse Linked List

Track three pointers: prev, curr, next.

```
Before: 1 -> 2 -> 3 -> None
After:  None <- 1 <- 2 <- 3
```

````python
def reverse_list(head):
    """Reverse linked list iteratively."""
    prev = None
    curr = head
    
    while curr:
        next_temp = curr.next  # Save next
        curr.next = prev       # Reverse link
        prev = curr            # Move prev
        curr = next_temp       # Move curr
    
    return prev

def reverse_recursive(head):
    """Reverse linked list recursively."""
    if not head or not head.next:
        return head
    
    new_head = reverse_recursive(head.next)
    head.next.next = head
    head.next = None
    
    return new_head

# Demo
head = build_list([1, 2, 3, 4, 5])
print(f"Original: {head}")
reversed_head = reverse_list(head)
print(f"Reversed: {reversed_head}")
````

### 4. Merge Two Lists

````python
def merge_sorted(l1, l2):
    """Merge two sorted linked lists."""
    dummy = ListNode()
    curr = dummy
    
    while l1 and l2:
        if l1.val <= l2.val:
            curr.next = l1
            l1 = l1.next
        else:
            curr.next = l2
            l2 = l2.next
        curr = curr.next
    
    curr.next = l1 or l2
    return dummy.next

# Demo
l1 = build_list([1, 3, 5])
l2 = build_list([2, 4, 6])
print(f"L1: {l1}")
print(f"L2: {l2}")
print(f"Merged: {merge_sorted(l1, l2)}")
````

## Real-World Use Cases

### 1. Undo/Redo Functionality
- Each action stored as node
- Undo = move backward, Redo = move forward
- Doubly linked list ideal

### 2. Browser History
- Back/Forward navigation
- Each page visit = new node
- Doubly linked list for bidirectional traversal

### 3. Music Playlist
- Songs as nodes
- Next/Previous song
- Circular list for repeat mode

### 4. LRU Cache (Least Recently Used)
- Doubly linked list + hash map
- O(1) access, insert, delete
- Used in caching systems, OS page replacement

### 5. Memory Allocation
- Free memory blocks as linked list
- malloc/free implementations

### 6. Polynomial Representation
- Each term as node (coefficient, exponent)
- Easy addition/multiplication

### 7. Graph Adjacency Lists
- Each vertex has linked list of neighbors
- Space-efficient for sparse graphs

### 8. Hash Table Chaining
- Collision resolution
- Each bucket is a linked list

````python
# Example: LRU Cache using OrderedDict (conceptually doubly linked list + hashmap)
from collections import OrderedDict

class LRUCache:
    def __init__(self, capacity: int):
        self.cache = OrderedDict()
        self.capacity = capacity
    
    def get(self, key: int) -> int:
        if key not in self.cache:
            return -1
        self.cache.move_to_end(key)  # Mark as recently used
        return self.cache[key]
    
    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            self.cache.move_to_end(key)
        self.cache[key] = value
        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)  # Remove least recently used

# Demo
lru = LRUCache(2)
lru.put(1, 1)
lru.put(2, 2)
print(f"Get 1: {lru.get(1)}")
lru.put(3, 3)  # Evicts key 2
print(f"Get 2: {lru.get(2)}")
print(f"Get 3: {lru.get(3)}")
````

---
## Key Tips

1. **Use dummy head** to simplify edge cases
2. **Draw the pointers** before coding - visualize connections
3. **Fast/slow pointers** for cycle and middle problems
4. **Save next before modifying** when reversing or deleting
5. **Consider edge cases**: empty list, single node, cycle
6. **Two-pass vs one-pass**: Sometimes two passes is clearer
7. **Recursion**: Elegant but uses O(n) stack space
