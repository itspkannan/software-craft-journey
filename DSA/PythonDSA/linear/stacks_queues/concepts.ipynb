# Stacks & Queues - Concepts & Use Cases

**Stacks** and **Queues** are fundamental linear data structures with restricted access patterns. Stacks follow LIFO (Last In, First Out), while queues follow FIFO (First In, First Out).

---
## Stack (LIFO)

```
        ┌───┐
push -> │ 4 │ <- pop (top)
        ├───┤
        │ 3 │
        ├───┤
        │ 2 │
        ├───┤
        │ 1 │
        └───┘

Last In, First Out
```

### Operations
| Operation | Description | Time |
|-----------|-------------|------|
| `push(x)` | Add element to top | O(1) |
| `pop()` | Remove and return top | O(1) |
| `peek()/top()` | View top without removing | O(1) |
| `isEmpty()` | Check if empty | O(1) |

````python
# Stack in Python - use list
stack = []

# Push
stack.append(1)
stack.append(2)
stack.append(3)
print(f"Stack after pushes: {stack}")

# Peek
print(f"Top element: {stack[-1]}")

# Pop
top = stack.pop()
print(f"Popped: {top}")
print(f"Stack after pop: {stack}")

# Check empty
print(f"Is empty: {len(stack) == 0}")
````

## Queue (FIFO)

```
                    front              rear
                      ↓                  ↓
dequeue <- ┌───┬───┬───┬───┐ <- enqueue
           │ 1 │ 2 │ 3 │ 4 │
           └───┴───┴───┴───┘

First In, First Out
```

### Operations
| Operation | Description | Time |
|-----------|-------------|------|
| `enqueue(x)` | Add to rear | O(1) |
| `dequeue()` | Remove from front | O(1) |
| `peek()/front()` | View front | O(1) |
| `isEmpty()` | Check if empty | O(1) |

````python
from collections import deque

# Queue in Python - use deque (NOT list!)
# list.pop(0) is O(n), deque.popleft() is O(1)
queue = deque()

# Enqueue
queue.append(1)
queue.append(2)
queue.append(3)
print(f"Queue after enqueues: {list(queue)}")

# Peek front
print(f"Front element: {queue[0]}")

# Dequeue
front = queue.popleft()
print(f"Dequeued: {front}")
print(f"Queue after dequeue: {list(queue)}")
````

## Deque (Double-Ended Queue)

```
      front                    rear
        ↓                        ↓
   ← ┌───┬───┬───┬───┐ →
     │ 1 │ 2 │ 3 │ 4 │
   → └───┴───┴───┴───┘ ←

Add/remove from both ends in O(1)
```

````python
from collections import deque

dq = deque([2, 3])
print(f"Initial: {list(dq)}")

# Add to both ends
dq.appendleft(1)  # Add to front
dq.append(4)      # Add to rear
print(f"After adds: {list(dq)}")

# Remove from both ends
front = dq.popleft()
rear = dq.pop()
print(f"Removed front={front}, rear={rear}")
print(f"After removes: {list(dq)}")
````

## Monotonic Stack

Stack that maintains elements in sorted order (increasing or decreasing).

### Monotonic Decreasing Stack
```
Elements: 3, 1, 4, 1, 5

Process 3: stack = [3]
Process 1: stack = [3, 1]     (1 < 3, push)
Process 4: stack = [4]        (4 > 1, 3, pop until can push)
Process 1: stack = [4, 1]
Process 5: stack = [5]
```

**Use cases**: Next greater element, largest rectangle in histogram

````python
def next_greater_element(nums):
    """Find next greater element for each position."""
    n = len(nums)
    result = [-1] * n
    stack = []  # Store indices
    
    for i in range(n):
        # Pop elements smaller than current
        while stack and nums[stack[-1]] < nums[i]:
            idx = stack.pop()
            result[idx] = nums[i]
        stack.append(i)
    
    return result

nums = [2, 1, 2, 4, 3]
print(f"Array: {nums}")
print(f"Next greater: {next_greater_element(nums)}")
````

## Monotonic Deque

Deque maintaining monotonic property. Used for sliding window max/min.

```
Array: [1, 3, -1, -3, 5, 3, 6, 7], k=3

Window [1,3,-1]: max=3 (deque stores index of 3)
Window [3,-1,-3]: max=3
Window [-1,-3,5]: max=5
...
```

````python
from collections import deque

def sliding_window_max(nums, k):
    """Find max in each sliding window of size k."""
    dq = deque()  # Store indices
    result = []
    
    for i, num in enumerate(nums):
        # Remove indices outside window
        while dq and dq[0] < i - k + 1:
            dq.popleft()
        
        # Remove smaller elements (they'll never be max)
        while dq and nums[dq[-1]] < num:
            dq.pop()
        
        dq.append(i)
        
        if i >= k - 1:
            result.append(nums[dq[0]])
    
    return result

nums = [1, 3, -1, -3, 5, 3, 6, 7]
print(f"Array: {nums}")
print(f"Window max (k=3): {sliding_window_max(nums, 3)}")
````

## Common Patterns

### 1. Matching Pairs (Stack)
- Valid parentheses
- HTML/XML tag matching

### 2. Expression Evaluation (Stack)
- Infix to postfix conversion
- Calculate postfix/prefix expressions

### 3. BFS Traversal (Queue)
- Level-order tree traversal
- Shortest path in unweighted graph

### 4. Sliding Window Max/Min (Monotonic Deque)
- Max/min in window of size k
- Jump game variations

````python
# Pattern: Valid Parentheses
def is_valid_parentheses(s):
    stack = []
    mapping = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in mapping:  # Closing bracket
            if not stack or stack.pop() != mapping[char]:
                return False
        else:  # Opening bracket
            stack.append(char)
    
    return len(stack) == 0

print(f"'()[]{{}}' valid: {is_valid_parentheses('()[]{}')}") 
print(f"'([)]' valid: {is_valid_parentheses('([)]}')}") 
````

````python
# Pattern: Evaluate Reverse Polish Notation
def eval_rpn(tokens):
    stack = []
    operators = {'+', '-', '*', '/'}
    
    for token in tokens:
        if token in operators:
            b, a = stack.pop(), stack.pop()
            if token == '+': result = a + b
            elif token == '-': result = a - b
            elif token == '*': result = a * b
            else: result = int(a / b)  # Truncate toward zero
            stack.append(result)
        else:
            stack.append(int(token))
    
    return stack[0]

# (2 + 1) * 3 = 9
tokens = ["2", "1", "+", "3", "*"]
print(f"{tokens} = {eval_rpn(tokens)}")
````

## Real-World Use Cases

### Stack Applications

1. **Function Call Stack**
   - Tracks function calls and returns
   - Local variables, return addresses
   - Stack overflow = too deep recursion

2. **Undo/Redo Operations**
   - Text editors, image editors
   - Each action pushed to stack
   - Undo = pop from undo stack, push to redo stack

3. **Browser Back Button**
   - Pages visited pushed to stack
   - Back = pop and display previous

4. **Syntax Parsing**
   - Compilers use stack for parsing
   - Expression evaluation
   - Bracket matching

5. **DFS Traversal**
   - Explicit stack instead of recursion
   - Tree/graph exploration

### Queue Applications

1. **Task Scheduling**
   - CPU scheduling (round-robin)
   - Print job spooling
   - Process management

2. **Message Queues**
   - RabbitMQ, Kafka, SQS
   - Async communication between services

3. **BFS Traversal**
   - Shortest path in unweighted graphs
   - Level-order tree traversal

4. **Request Handling**
   - Web server request queues
   - Load balancing

5. **Buffering**
   - IO buffers
   - Video streaming buffers
   - Keyboard input buffer

````python
# Example: Simple Task Queue
from collections import deque
import time

class TaskQueue:
    def __init__(self):
        self.queue = deque()
    
    def add_task(self, task_name):
        self.queue.append(task_name)
        print(f"Added task: {task_name}")
    
    def process_next(self):
        if self.queue:
            task = self.queue.popleft()
            print(f"Processing: {task}")
            return task
        print("No tasks in queue")
        return None

# Demo
tq = TaskQueue()
tq.add_task("Send email")
tq.add_task("Generate report")
tq.add_task("Update database")

print("\n--- Processing ---")
while tq.queue:
    tq.process_next()
````

---
## Key Tips

1. **Use deque for queue** - `list.pop(0)` is O(n), `deque.popleft()` is O(1)
2. **Monotonic stack** for "next greater/smaller" problems
3. **Stack for matching pairs** - parentheses, tags, brackets
4. **Queue for BFS** - level-order, shortest path
5. **Two stacks = queue** - amortized O(1) operations
6. **Check empty before pop** - avoid IndexError
7. **Store indices in stack** when you need positions, not just values
