## Q1: Binary Search

### Approach: Standard Binary Search
- Compare middle element with target
- Narrow search to left or right half

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```python
def search(nums: List[int], target: int) -> int:
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1

print("Solution defined!")
```

## Q2: Search Insert Position

### Approach: Binary Search for Lower Bound
- Find leftmost position where target could be inserted

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```python
def search_insert(nums: List[int], target: int) -> int:
    left, right = 0, len(nums)
    
    while left < right:
        mid = left + (right - left) // 2
        
        if nums[mid] < target:
            left = mid + 1
        else:
            right = mid
    
    return left

print("Solution defined!")
```

## Q3: Find First and Last Position

### Approach: Two Binary Searches
- Find leftmost occurrence
- Find rightmost occurrence

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```python
def search_range(nums: List[int], target: int) -> List[int]:
    def find_left():
        left, right = 0, len(nums)
        while left < right:
            mid = (left + right) // 2
            if nums[mid] < target:
                left = mid + 1
            else:
                right = mid
        return left
    
    def find_right():
        left, right = 0, len(nums)
        while left < right:
            mid = (left + right) // 2
            if nums[mid] <= target:
                left = mid + 1
            else:
                right = mid
        return left - 1
    
    left_idx = find_left()
    
    if left_idx >= len(nums) or nums[left_idx] != target:
        return [-1, -1]
    
    return [left_idx, find_right()]

print("Solution defined!")
```

## Q4: Search in Rotated Sorted Array

### Approach: Modified Binary Search
- One half is always sorted
- Check if target is in sorted half

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```python
def search_rotated(nums: List[int], target: int) -> int:
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if nums[mid] == target:
            return mid
        
        # Left half is sorted
        if nums[left] <= nums[mid]:
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        # Right half is sorted
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1
    
    return -1

print("Solution defined!")
```

## Q5: Find Minimum in Rotated Sorted Array

### Approach: Binary Search for Pivot
- Compare mid with right to determine which half has minimum

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```python
def find_min(nums: List[int]) -> int:
    left, right = 0, len(nums) - 1
    
    while left < right:
        mid = (left + right) // 2
        
        # Mid is greater than right -> min is in right half
        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid
    
    return nums[left]

print("Solution defined!")
```

## Q6: Find Peak Element

### Approach: Binary Search
- If mid > mid+1, peak is on left (including mid)
- Else peak is on right

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```python
def find_peak_element(nums: List[int]) -> int:
    left, right = 0, len(nums) - 1
    
    while left < right:
        mid = (left + right) // 2
        
        if nums[mid] > nums[mid + 1]:
            right = mid  # Peak on left (including mid)
        else:
            left = mid + 1  # Peak on right
    
    return left

print("Solution defined!")
```

## Q7: Koko Eating Bananas

### Approach: Binary Search on Answer
- Search space: [1, max(piles)]
- For each speed k, check if can finish in h hours

### Complexity
- **Time:** O(n log m) where m = max(piles)
- **Space:** O(1)

```python
def min_eating_speed(piles: List[int], h: int) -> int:
    def can_finish(k):
        hours = sum(math.ceil(pile / k) for pile in piles)
        return hours <= h
    
    left, right = 1, max(piles)
    
    while left < right:
        mid = (left + right) // 2
        
        if can_finish(mid):
            right = mid
        else:
            left = mid + 1
    
    return left

print("Solution defined!")
```

## Q8: Search a 2D Matrix

### Approach: Treat as 1D Array
- Treat matrix as 1D sorted array
- Convert index: row = idx // cols, col = idx % cols

### Complexity
- **Time:** O(log(m × n))
- **Space:** O(1)

```python
def search_matrix(matrix: List[List[int]], target: int) -> bool:
    if not matrix:
        return False
    
    rows, cols = len(matrix), len(matrix[0])
    left, right = 0, rows * cols - 1
    
    while left <= right:
        mid = (left + right) // 2
        row, col = mid // cols, mid % cols
        val = matrix[row][col]
        
        if val == target:
            return True
        elif val < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return False

print("Solution defined!")
```
