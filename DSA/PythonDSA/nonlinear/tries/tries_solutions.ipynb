## Q1: Implement Trie

### Approach: Nested Dictionary / TrieNode
- Each node has children dict and is_end flag
- Follow/create path for insert
- Follow path for search/startsWith

### Complexity
- **Time:** O(m) for all operations (m = word length)
- **Space:** O(sum of all word lengths)

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class Trie:
    def __init__(self):
        self.root = TrieNode()
    
    def insert(self, word: str) -> None:
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True
    
    def search(self, word: str) -> bool:
        node = self._find_prefix(word)
        return node is not None and node.is_end
    
    def startsWith(self, prefix: str) -> bool:
        return self._find_prefix(prefix) is not None
    
    def _find_prefix(self, prefix: str):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return None
            node = node.children[char]
        return node

print("Solution defined!")
```

## Q2: Add and Search Words

### Approach: Trie + DFS for Wildcards
- Standard trie insert
- DFS search with branching on '.'

### Complexity
- **Time:** O(m) insert, O(26^m) worst case search
- **Space:** O(sum of all word lengths)

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False

class WordDictionary:
    def __init__(self):
        self.root = TrieNode()
    
    def addWord(self, word: str) -> None:
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True
    
    def search(self, word: str) -> bool:
        def dfs(node, i):
            if i == len(word):
                return node.is_end
            
            char = word[i]
            
            if char == '.':
                # Try all children
                for child in node.children.values():
                    if dfs(child, i + 1):
                        return True
                return False
            else:
                if char not in node.children:
                    return False
                return dfs(node.children[char], i + 1)
        
        return dfs(self.root, 0)

print("Solution defined!")
```

## Q3: Word Search II

### Approach: Trie + Board DFS
- Build trie from words
- DFS from each cell, pruning with trie

### Complexity
- **Time:** O(m × n × 4^L) where L = max word length
- **Space:** O(sum of word lengths)

```python
def find_words(board: List[List[str]], words: List[str]) -> List[str]:
    # Build trie
    trie = {}
    for word in words:
        node = trie
        for char in word:
            node = node.setdefault(char, {})
        node['$'] = word  # Mark end with the word itself
    
    rows, cols = len(board), len(board[0])
    result = []
    
    def dfs(r, c, node):
        char = board[r][c]
        
        if char not in node:
            return
        
        next_node = node[char]
        
        # Found a word
        if '$' in next_node:
            result.append(next_node['$'])
            del next_node['$']  # Avoid duplicates
        
        # Mark visited
        board[r][c] = '#'
        
        # Explore neighbors
        for dr, dc in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols and board[nr][nc] != '#':
                dfs(nr, nc, next_node)
        
        # Restore
        board[r][c] = char
    
    for r in range(rows):
        for c in range(cols):
            dfs(r, c, trie)
    
    return result

print("Solution defined!")
```

## Q4: Replace Words

### Approach: Trie for Root Lookup
- Build trie from dictionary
- For each word, find shortest root

### Complexity
- **Time:** O(d + s) where d = dict size, s = sentence size
- **Space:** O(d)

```python
def replace_words(dictionary: List[str], sentence: str) -> str:
    # Build trie
    trie = {}
    for root in dictionary:
        node = trie
        for char in root:
            node = node.setdefault(char, {})
        node['$'] = root  # Store the root word
    
    def find_root(word):
        node = trie
        for char in word:
            if '$' in node:  # Found shortest root
                return node['$']
            if char not in node:
                return word
            node = node[char]
        return node.get('$', word)
    
    return ' '.join(find_root(word) for word in sentence.split())

print("Solution defined!")
```

## Q5: Longest Word in Dictionary

### Approach: Trie + DFS
- Build trie marking end of words
- DFS only through nodes that are word endings

### Complexity
- **Time:** O(sum of word lengths)
- **Space:** O(sum of word lengths)

```python
def longest_word(words: List[str]) -> str:
    # Build trie
    trie = {}
    for word in words:
        node = trie
        for char in word:
            node = node.setdefault(char, {})
        node['$'] = True
    
    result = ""
    
    def dfs(node, path):
        nonlocal result
        
        for char in sorted(node.keys()):  # Lexicographic order
            if char == '$':
                continue
            
            child = node[char]
            if '$' in child:  # Only proceed if it's a word ending
                new_path = path + char
                
                if len(new_path) > len(result):
                    result = new_path
                
                dfs(child, new_path)
    
    dfs(trie, "")
    return result

print("Solution defined!")
```

## Q6: Magic Dictionary

### Approach: Trie + DFS with One Change
- Build trie from dictionary
- DFS allowing exactly one character mismatch

### Complexity
- **Time:** O(m × 26) per search
- **Space:** O(dictionary size)

```python
class MagicDictionary:
    def __init__(self):
        self.trie = {}
    
    def buildDict(self, dictionary: List[str]) -> None:
        for word in dictionary:
            node = self.trie
            for char in word:
                node = node.setdefault(char, {})
            node['$'] = True
    
    def search(self, searchWord: str) -> bool:
        def dfs(node, i, changed):
            if i == len(searchWord):
                return '$' in node and changed
            
            char = searchWord[i]
            
            for c, child in node.items():
                if c == '$':
                    continue
                
                if c == char:
                    if dfs(child, i + 1, changed):
                        return True
                elif not changed:  # Use the one change
                    if dfs(child, i + 1, True):
                        return True
            
            return False
        
        return dfs(self.trie, 0, False)

print("Solution defined!")
```

## Q7: Stream of Characters

### Approach: Reversed Trie
- Store words reversed in trie
- Keep stream buffer
- Search backwards in stream

### Complexity
- **Time:** O(max word length) per query
- **Space:** O(sum of word lengths + stream length)

```python
class StreamChecker:
    def __init__(self, words: List[str]):
        # Build reversed trie
        self.trie = {}
        self.max_len = 0
        
        for word in words:
            self.max_len = max(self.max_len, len(word))
            node = self.trie
            for char in reversed(word):
                node = node.setdefault(char, {})
            node['$'] = True
        
        self.stream = []
    
    def query(self, letter: str) -> bool:
        self.stream.append(letter)
        
        node = self.trie
        
        # Search backwards through stream
        for i in range(len(self.stream) - 1, max(-1, len(self.stream) - self.max_len - 1), -1):
            char = self.stream[i]
            
            if char not in node:
                return False
            
            node = node[char]
            
            if '$' in node:
                return True
        
        return False

print("Solution defined!")
```
