# Trees - Concepts & Use Cases

A **tree** is a hierarchical data structure consisting of nodes connected by edges, with a single root node and no cycles.

---
## Core Terminology

```
           1          <- Root (depth 0, height 3)
          / \
         2   3        <- Internal nodes (depth 1)
        / \   \
       4   5   6      <- Leaves (depth 2)
      /
     7                <- Leaf (depth 3)
```

| Term | Definition |
|------|------------|
| **Root** | Top node with no parent |
| **Leaf** | Node with no children |
| **Internal Node** | Node with at least one child |
| **Depth** | Distance from root to node |
| **Height** | Max distance from node to any leaf |
| **Subtree** | Tree formed by a node and all descendants |
| **Sibling** | Nodes sharing the same parent |

---
## Types of Trees

### Binary Tree
Each node has at most 2 children (left and right).

```
       1
      / \
     2   3
    / \
   4   5
```

### Binary Search Tree (BST)
Binary tree where: `left < parent < right`

```
       8
      / \
     3   10
    / \    \
   1   6    14
      / \
     4   7
```

### Balanced Trees
- **AVL Tree**: Height difference between subtrees ≤ 1
- **Red-Black Tree**: Self-balancing with color properties
- **B-Tree**: Multi-way tree for databases/filesystems

### Special Trees
- **Complete Binary Tree**: All levels filled except possibly last (filled left to right)
- **Perfect Binary Tree**: All internal nodes have 2 children, all leaves at same level
- **Full Binary Tree**: Every node has 0 or 2 children

## Tree Traversals

```
       1
      / \
     2   3
    / \
   4   5
```

| Traversal | Order | Result | Use Case |
|-----------|-------|--------|----------|
| **Preorder** | Root → Left → Right | 1,2,4,5,3 | Copy tree, serialize |
| **Inorder** | Left → Root → Right | 4,2,5,1,3 | BST sorted order |
| **Postorder** | Left → Right → Root | 4,5,2,3,1 | Delete tree, eval expression |
| **Level Order** | Level by level | 1,2,3,4,5 | BFS, shortest path |

````python
from typing import Optional, List
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def preorder(root: Optional[TreeNode]) -> List[int]:
    """Root -> Left -> Right"""
    if not root:
        return []
    return [root.val] + preorder(root.left) + preorder(root.right)

def inorder(root: Optional[TreeNode]) -> List[int]:
    """Left -> Root -> Right (sorted for BST)"""
    if not root:
        return []
    return inorder(root.left) + [root.val] + inorder(root.right)

def postorder(root: Optional[TreeNode]) -> List[int]:
    """Left -> Right -> Root"""
    if not root:
        return []
    return postorder(root.left) + postorder(root.right) + [root.val]

def level_order(root: Optional[TreeNode]) -> List[List[int]]:
    """BFS level by level"""
    if not root:
        return []
    result, queue = [], deque([root])
    while queue:
        level = []
        for _ in range(len(queue)):
            node = queue.popleft()
            level.append(node.val)
            if node.left: queue.append(node.left)
            if node.right: queue.append(node.right)
        result.append(level)
    return result

# Demo
root = TreeNode(1, TreeNode(2, TreeNode(4), TreeNode(5)), TreeNode(3))
print(f"Preorder:  {preorder(root)}")
print(f"Inorder:   {inorder(root)}")
print(f"Postorder: {postorder(root)}")
print(f"Level:     {level_order(root)}")
````

## Time & Space Complexity

### Binary Search Tree (BST)

| Operation | Average | Worst (Skewed) |
|-----------|---------|----------------|
| Search | O(log n) | O(n) |
| Insert | O(log n) | O(n) |
| Delete | O(log n) | O(n) |

### Balanced Trees (AVL, Red-Black)

| Operation | Time |
|-----------|------|
| Search | O(log n) |
| Insert | O(log n) |
| Delete | O(log n) |

### Space Complexity
- **Storage**: O(n)
- **Recursive traversal**: O(h) where h = height

## Real-World Use Cases

### 1. File Systems
Directory structure is a tree where folders are internal nodes and files are leaves.

```
/
├── home/
│   ├── user/
│   │   ├── documents/
│   │   └── downloads/
│   └── admin/
└── etc/
    └── config.txt
```

### 2. Database Indexing (B-Trees)
Databases use B-Trees/B+ Trees for efficient indexing:
- MySQL InnoDB uses B+ Trees
- PostgreSQL uses B-Trees for indexes
- Enables O(log n) lookups on millions of records

### 3. DOM (Document Object Model)
HTML/XML documents are parsed into tree structures:

```html
<html>           <!-- Root -->
  <head>         <!-- Child -->
    <title>      <!-- Grandchild -->
  <body>
    <div>
      <p>
```

### 4. Expression Evaluation
Compilers use Abstract Syntax Trees (AST) to parse expressions:

```
Expression: (3 + 4) * 5

      *
     / \
    +   5
   / \
  3   4
```

### 5. Decision Trees (ML)
Machine learning uses trees for classification:

```
              [Age > 30?]
              /         \
           Yes           No
            |             |
      [Income > 50k?]   [Student?]
         /    \          /    \
       Yes    No       Yes    No
        |      |        |      |
      Buy   Don't    Don't   Buy
```

### 6. Network Routing
Spanning trees in network switches prevent loops:
- **Spanning Tree Protocol (STP)**: Creates loop-free topology
- **Routing tables**: Hierarchical routing

### 7. Game AI (Minimax Trees)
Chess/Tic-tac-toe AI uses game trees:
- Each node = game state
- Edges = possible moves
- Minimax algorithm traverses to find optimal move

### 8. Org Charts / Hierarchies
Company structures, taxonomies, family trees:

```
           CEO
          / | \
       CTO CFO COO
       /|
    Dev QA
```

def dfs(node):
    if not node:

### 2. BFS Level Order Template
```python
def bfs(root):
    queue = deque([root])
    while queue:
        node = queue.popleft()
        process(node)
        if node.left: queue.append(node.left)
        if node.right: queue.append(node.right)
```

### 3. Path Problems
- Pass path down via parameters
- Return results up via return values

### 4. Tree Construction
- Build from traversal orders (preorder + inorder)
- Recursive divide and conquer

````python
# Interactive: Visualize a BST
class BST:
    def __init__(self):
        self.root = None
    
    def insert(self, val):
        if not self.root:
            self.root = TreeNode(val)
            return
        
        curr = self.root
        while True:
            if val < curr.val:
                if not curr.left:
                    curr.left = TreeNode(val)
                    return
                curr = curr.left
            else:
                if not curr.right:
                    curr.right = TreeNode(val)
                    return
                curr = curr.right
    
    def display(self, node=None, level=0, prefix="Root: "):
        if node is None:
            node = self.root
        if node is not None:
            print(" " * (level * 4) + prefix + str(node.val))
            if node.left or node.right:
                if node.left:
                    self.display(node.left, level + 1, "L--- ")
                else:
                    print(" " * ((level + 1) * 4) + "L--- None")
                if node.right:
                    self.display(node.right, level + 1, "R--- ")
                else:
                    print(" " * ((level + 1) * 4) + "R--- None")

# Build and visualize
bst = BST()
for val in [8, 3, 10, 1, 6, 14, 4, 7]:
    bst.insert(val)

print("BST Structure:")
bst.display()
print(f"\nInorder (sorted): {inorder(bst.root)}")
````

## Key Tips

1. **Always handle null/empty tree** as base case
2. **Think recursively** - most tree problems have elegant recursive solutions
3. **Identify traversal type needed**:
   - Need sorted order? → Inorder
   - Need level-by-level? → BFS
   - Need to process children first? → Postorder
4. **For BST problems**, leverage the ordering property
5. **Path problems** often need to track state while traversing
