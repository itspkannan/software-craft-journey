## Q1: Valid Parentheses

### Approach: Stack
- Push opening brackets
- For closing brackets, check if matches top

### Complexity
- **Time:** O(n)
- **Space:** O(n)

```python
def is_valid(s: str) -> bool:
    stack = []
    mapping = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in mapping:
            if not stack or stack.pop() != mapping[char]:
                return False
        else:
            stack.append(char)
    
    return len(stack) == 0

print("Solution defined!")
```

## Q2: Min Stack

### Approach: Two Stacks
- Main stack for values
- Min stack tracks minimum at each level

### Complexity
- **Time:** O(1) for all operations
- **Space:** O(n)

```python
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []
    
    def push(self, val: int) -> None:
        self.stack.append(val)
        # Push min so far
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)
    
    def pop(self) -> None:
        val = self.stack.pop()
        if val == self.min_stack[-1]:
            self.min_stack.pop()
    
    def top(self) -> int:
        return self.stack[-1]
    
    def getMin(self) -> int:
        return self.min_stack[-1]

print("Solution defined!")
```

## Q3: Daily Temperatures

### Approach: Monotonic Stack (Decreasing)
- Stack stores indices of unresolved temperatures
- When we find warmer temp, pop and calculate days

### Complexity
- **Time:** O(n)
- **Space:** O(n)

```python
def daily_temperatures(temperatures: List[int]) -> List[int]:
    n = len(temperatures)
    result = [0] * n
    stack = []  # Store indices
    
    for i, temp in enumerate(temperatures):
        # Pop all indices with lower temp
        while stack and temp > temperatures[stack[-1]]:
            idx = stack.pop()
            result[idx] = i - idx
        stack.append(i)
    
    return result

print("Solution defined!")
```

## Q4: Evaluate Reverse Polish Notation

### Approach: Stack
- Push numbers
- On operator, pop two operands, compute, push result

### Complexity
- **Time:** O(n)
- **Space:** O(n)

```python
def eval_rpn(tokens: List[str]) -> int:
    stack = []
    operators = {'+', '-', '*', '/'}
    
    for token in tokens:
        if token in operators:
            b = stack.pop()
            a = stack.pop()
            
            if token == '+':
                stack.append(a + b)
            elif token == '-':
                stack.append(a - b)
            elif token == '*':
                stack.append(a * b)
            else:  # Division truncates toward zero
                stack.append(int(a / b))
        else:
            stack.append(int(token))
    
    return stack[0]

print("Solution defined!")
```

## Q5: Largest Rectangle in Histogram

### Approach: Monotonic Stack (Increasing)
- Maintain increasing stack of indices
- When smaller bar found, calculate area for popped bars

### Complexity
- **Time:** O(n)
- **Space:** O(n)

```python
def largest_rectangle_area(heights: List[int]) -> int:
    stack = []  # Indices
    max_area = 0
    heights = [0] + heights + [0]  # Sentinel values
    
    for i, h in enumerate(heights):
        while stack and heights[stack[-1]] > h:
            height = heights[stack.pop()]
            width = i - stack[-1] - 1
            max_area = max(max_area, height * width)
        stack.append(i)
    
    return max_area

print("Solution defined!")
```

## Q6: Implement Queue using Stacks

### Approach: Two Stacks (Lazy Transfer)
- in_stack for push
- out_stack for pop/peek
- Transfer when out_stack empty

### Complexity
- **Time:** Amortized O(1) for all operations
- **Space:** O(n)

```python
class MyQueue:
    def __init__(self):
        self.in_stack = []
        self.out_stack = []
    
    def push(self, x: int) -> None:
        self.in_stack.append(x)
    
    def pop(self) -> int:
        self._transfer()
        return self.out_stack.pop()
    
    def peek(self) -> int:
        self._transfer()
        return self.out_stack[-1]
    
    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack
    
    def _transfer(self):
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())

print("Solution defined!")
```

## Q7: Sliding Window Maximum

### Approach: Monotonic Deque (Decreasing)
- Deque stores indices in decreasing order of values
- Front of deque is always the max

### Complexity
- **Time:** O(n)
- **Space:** O(k)

```python
def max_sliding_window(nums: List[int], k: int) -> List[int]:
    dq = deque()  # Stores indices
    result = []
    
    for i in range(len(nums)):
        # Remove indices outside window
        while dq and dq[0] < i - k + 1:
            dq.popleft()
        
        # Remove smaller elements (they'll never be max)
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
        
        dq.append(i)
        
        # Add to result once window is full
        if i >= k - 1:
            result.append(nums[dq[0]])
    
    return result

print("Solution defined!")
```

## Q8: Next Greater Element I

### Approach: Monotonic Stack + Hash Map
- Build next greater map for nums2
- Look up each nums1 element

### Complexity
- **Time:** O(n + m)
- **Space:** O(n)

```python
def next_greater_element(nums1: List[int], nums2: List[int]) -> List[int]:
    next_greater = {}  # num -> next greater
    stack = []
    
    for num in nums2:
        while stack and num > stack[-1]:
            next_greater[stack.pop()] = num
        stack.append(num)
    
    return [next_greater.get(num, -1) for num in nums1]

print("Solution defined!")
```
