# Binary Search - Concepts & Use Cases

**Binary Search** is a divide-and-conquer algorithm for finding elements in sorted arrays with O(log n) time complexity. It's one of the most important algorithmic patterns.

---
## Core Concept

Repeatedly divide search space in half by comparing with middle element.

```
Find 7 in: [1, 3, 5, 7, 9, 11, 13]

Step 1: [1, 3, 5, 7, 9, 11, 13]
 ↑
 mid=7 Found!

Find 11:
Step 1: [1, 3, 5, 7, 9, 11, 13] mid=7, 11>7, go right
Step 2: [9, 11, 13] mid=11, Found!
```

### Key Requirements
- **Sorted** array (or monotonic property)
- **Random access** (O(1) index access)

## Common Templates

### Template 1: Exact Match (left <= right)
```
while left <= right:
 mid = left + (right - left) // 2
 if nums[mid] == target:
 return mid
 elif nums[mid] < target:
 left = mid + 1
 else:
 right = mid - 1
return -1
```

### Template 2: Left Boundary (left < right)
```
while left < right:
 mid = left + (right - left) // 2
 if condition(mid):
 right = mid # Answer could be mid
 else:
 left = mid + 1 # Answer must be after mid
return left
```

### Template 3: Right Boundary
```
while left < right:
 mid = left + (right - left + 1) // 2 # Round up!
 if condition(mid):
 left = mid # Answer could be mid
 else:
 right = mid - 1 # Answer must be before mid
return left
```

## Common Mistakes

### 1. Integer Overflow
```
# Bad (can overflow in other languages)
mid = (left + right) / 2

# Good
mid = left + (right - left) // 2
```

### 2. Infinite Loop
```
# Bad: left never increases
while left < right:
 mid = (left + right) // 2
 if condition:
 left = mid # Should be mid + 1 or use ceiling division
```

### 3. Off-by-One Errors
- Using `<=` vs `<` in while condition
- Using `mid` vs `mid + 1` or `mid - 1`

### 4. Wrong Boundary Update
- `right = mid - 1` vs `right = mid`
- Depends on whether mid could be answer

---
## Key Tips

1. **Check if array is sorted** or has monotonic property
2. **Identify what you're searching for**: exact value, boundary, or answer
3. **Choose correct template** based on problem type
4. **Use `left + (right - left) // 2`** to avoid overflow
5. **Test edge cases**: empty array, single element, all same, target not found
6. **"Binary search on answer"** when searching in range of possible answers
7. **Round up for right boundary** to avoid infinite loop
