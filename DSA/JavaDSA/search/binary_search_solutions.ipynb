## Q1: Binary Search

### Approach: Standard Binary Search
- Compare middle element with target
- Narrow search to left or right half

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```java
/*
 * --- Reference solution (notation) ---
 * def search(nums: List[int], target: int) -> int:
 *  left, right = 0, len(nums) - 1
 *  
 *  while left <= right:
 *  mid = left + (right - left) // 2
 *  
 *  if nums[mid] == target:
 *  return mid
 *  elif nums[mid] < target:
 *  left = mid + 1
 *  else:
 *  right = mid - 1
 *  
 *  return -1
 * 
 * print("Solution defined!")
 */

static int search(int[] nums, int target) {
        var left = 0;
        var right = (nums.length - 1);
        while (((left <= right))) {
            var mid = (left + ((right - left) / 2));
            if (((nums[mid] == target))) {
                return mid;
            } else {
                if (((nums[mid] < target))) {
                    left = (mid + 1);
                } else {
                    right = (mid - 1);
                }
            }
        }
        return (-(1));
}
System.out.println("All tests passed!");
```

## Q2: Search Insert Position

### Approach: Binary Search for Lower Bound
- Find leftmost position where target could be inserted

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```java
/*
 * --- Reference solution (notation) ---
 * def search_insert(nums: List[int], target: int) -> int:
 *  left, right = 0, len(nums)
 *  
 *  while left < right:
 *  mid = left + (right - left) // 2
 *  
 *  if nums[mid] < target:
 *  left = mid + 1
 *  else:
 *  right = mid
 *  
 *  return left
 * 
 * print("Solution defined!")
 */

static int searchInsert(int[] nums, int target) {
        var left = 0;
        var right = nums.length;
        while (((left < right))) {
            var mid = (left + ((right - left) / 2));
            if (((nums[mid] < target))) {
                left = (mid + 1);
            } else {
                right = mid;
            }
        }
        return left;
}
System.out.println("All tests passed!");
```

## Q4: Search in Rotated Sorted Array

### Approach: Modified Binary Search
- One half is always sorted
- Check if target is in sorted half

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```java
/*
 * --- Reference solution (notation) ---
 * def search_rotated(nums: List[int], target: int) -> int:
 *  left, right = 0, len(nums) - 1
 *  
 *  while left <= right:
 *  mid = (left + right) // 2
 *  
 *  if nums[mid] == target:
 *  return mid
 *  
 *  # Left half is sorted
 *  if nums[left] <= nums[mid]:
 *  if nums[left] <= target < nums[mid]:
 *  right = mid - 1
 *  else:
 *  left = mid + 1
 *  # Right half is sorted
 *  else:
 *  if nums[mid] < target <= nums[right]:
 *  left = mid + 1
 *  else:
 *  right = mid - 1
 *  
 *  return -1
 * 
 * print("Solution defined!")
 */

static int searchRotated(int[] nums, int target) {
        var left = 0;
        var right = (nums.length - 1);
        while (((left <= right))) {
            var mid = ((left + right) / 2);
            if (((nums[mid] == target))) {
                return mid;
            }
            if (((nums[left] <= nums[mid]))) {
                if (((nums[left] <= target) && (target < nums[mid]))) {
                    right = (mid - 1);
                } else {
                    left = (mid + 1);
                }
            } else {
                if (((nums[mid] < target) && (target <= nums[right]))) {
                    left = (mid + 1);
                } else {
                    right = (mid - 1);
                }
            }
        }
        return (-(1));
}
System.out.println("All tests passed!");
```

## Q5: Find Minimum in Rotated Sorted Array

### Approach: Binary Search for Pivot
- Compare mid with right to determine which half has minimum

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```java
/*
 * --- Reference solution (notation) ---
 * def find_min(nums: List[int]) -> int:
 *  left, right = 0, len(nums) - 1
 *  
 *  while left < right:
 *  mid = (left + right) // 2
 *  
 *  # Mid is greater than right -> min is in right half
 *  if nums[mid] > nums[right]:
 *  left = mid + 1
 *  else:
 *  right = mid
 *  
 *  return nums[left]
 * 
 * print("Solution defined!")
 */

static int findMin(int[] nums) {
        var left = 0;
        var right = (nums.length - 1);
        while (((left < right))) {
            var mid = ((left + right) / 2);
            if (((nums[mid] > nums[right]))) {
                left = (mid + 1);
            } else {
                right = mid;
            }
        }
        return nums[left];
}
System.out.println("All tests passed!");
```

## Q6: Find Peak Element

### Approach: Binary Search
- If mid > mid+1, peak is on left (including mid)
- Else peak is on right

### Complexity
- **Time:** O(log n)
- **Space:** O(1)

```java
/*
 * --- Reference solution (notation) ---
 * def find_peak_element(nums: List[int]) -> int:
 *  left, right = 0, len(nums) - 1
 *  
 *  while left < right:
 *  mid = (left + right) // 2
 *  
 *  if nums[mid] > nums[mid + 1]:
 *  right = mid # Peak on left (including mid)
 *  else:
 *  left = mid + 1 # Peak on right
 *  
 *  return left
 * 
 * print("Solution defined!")
 */

static int findPeakElement(int[] nums) {
        var left = 0;
        var right = (nums.length - 1);
        while (((left < right))) {
            var mid = ((left + right) / 2);
            if (((nums[mid] > nums[(mid + 1)]))) {
                right = mid;
            } else {
                left = (mid + 1);
            }
        }
        return left;
}
System.out.println("All tests passed!");
```
