# Linked Lists - Concepts & Use Cases

A **linked list** is a linear data structure where elements are stored in nodes, each containing data and a reference (pointer) to the next node.

---
## Core Concepts

### Singly Linked List
```
head
 ↓
+------+ +------+ +------+ +------+
| 1 | -> | 2 | -> | 3 | -> | 4 | -> None
+------+ +------+ +------+ +------+
```

### Doubly Linked List
```
head tail
 ↓ ↓
+------+ +------+ +------+ +------+
| 1 | ⇄ | 2 | ⇄ | 3 | ⇄ | 4 |
+------+ +------+ +------+ +------+
```

### Circular Linked List
```
 +------+ +------+ +------+
┌-> | 1 | -> | 2 | -> | 3 | -┐
| +------+ +------+ +------+ |
└─────────────────────────────────────┘
```

## Array vs Linked List

| Operation | Array | Linked List |
|-----------|-------|-------------|
| Access by index | O(1) | O(n) |
| Search | O(n) | O(n) |
| Insert at head | O(n) | O(1) |
| Insert at tail | O(1)* | O(n) or O(1)** |
| Insert at middle | O(n) | O(1)*** |
| Delete at head | O(n) | O(1) |
| Delete at tail | O(1) | O(n) or O(1)** |
| Memory | Contiguous | Scattered |
| Cache locality | Good | Poor |

*Amortized for dynamic arrays 
**With tail pointer 
***If you have reference to the node

## Real-World Use Cases

### 1. Undo/Redo Functionality
- Each action stored as node
- Undo = move backward, Redo = move forward
- Doubly linked list ideal

### 2. Browser History
- Back/Forward navigation
- Each page visit = new node
- Doubly linked list for bidirectional traversal

### 3. Music Playlist
- Songs as nodes
- Next/Previous song
- Circular list for repeat mode

### 4. LRU Cache (Least Recently Used)
- Doubly linked list + hash map
- O(1) access, insert, delete
- Used in caching systems, OS page replacement

### 5. Memory Allocation
- Free memory blocks as linked list
- malloc/free implementations

### 6. Polynomial Representation
- Each term as node (coefficient, exponent)
- Easy addition/multiplication

### 7. Graph Adjacency Lists
- Each vertex has linked list of neighbors
- Space-efficient for sparse graphs

### 8. Hash Table Chaining
- Collision resolution
- Each bucket is a linked list

## Key Tips

1. **Use dummy head** to simplify edge cases
2. **Draw the pointers** before coding - visualize connections
3. **Fast/slow pointers** for cycle and middle problems
4. **Save next before modifying** when reversing or deleting
5. **Consider edge cases**: empty list, single node, cycle
6. **Two-pass vs one-pass**: Sometimes two passes is clearer
7. **Recursion**: Elegant but uses O(n) stack space
