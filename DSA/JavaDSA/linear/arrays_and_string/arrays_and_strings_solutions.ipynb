## Q6: Maximum Subarray (Kadane's Algorithm)

**Problem:** Find contiguous subarray with largest sum.

### Approach: Kadane's Algorithm
- Track current sum; if it goes negative, reset to 0
- At each step, decide: extend current subarray or start fresh
- `current_sum = max(num, current_sum + num)`

### Complexity
- **Time:** O(n) - single pass
- **Space:** O(1) - two variables

```java
/*
 * --- Original exercise specification (reference) ---
 * def max_subarray(nums: list[int]) -> int:
 *  current_sum = nums[0]
 *  max_sum = nums[0]
 *  
 *  for num in nums[1:]:
 *  # Either extend current subarray or start new
 *  current_sum = max(num, current_sum + num)
 *  max_sum = max(max_sum, current_sum)
 *  
 *  return max_sum
 * 
 * # Test Cases
 * assert max_subarray([-2, 1, -3, 4, -1, 2, 1, -5, 4]) == 6
 * assert max_subarray([1]) == 1
 * assert max_subarray([5, 4, -1, 7, 8]) == 23
 * print("All tests passed!")
 */

static int maxSubarray(int[] nums) {
    // TODO: implement from specification above
}
T.eq("assert failed", maxSubarray(Arrays.asList(-2, 1, -3, 4, -1, 2, 1, -5, 4)), 6);
T.eq("assert failed", maxSubarray(new int[] {1}), 1);
T.eq("assert failed", maxSubarray(Arrays.asList(5, 4, -1, 7, 8)), 23);
System.out.println("All tests passed!");
```
