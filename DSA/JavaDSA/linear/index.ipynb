## Arrays & Strings

**Directory:** `arrays_and_string/`

The most fundamental data structures. Arrays store elements in contiguous memory with O(1) random access. Strings are character sequences with special operations.

### Key Concepts
- Contiguous memory, random access O(1)
- Two pointers technique
- Sliding window pattern
- Prefix sum for range queries
- Hash map for O(1) lookups

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Two Sum | Easy | Hash map |
| Valid Palindrome | Easy | Two pointers |
| Longest Substring | Medium | Sliding window |
| Container With Water | Medium | Two pointers |
| Group Anagrams | Medium | Sorting / Hash |

### Real-World Applications
- Image processing (2D arrays)
- Text processing and parsing
- Time series data
- Databases and indexing

## Linked Lists

**Directory:** `linked_list/`

Dynamic data structure with nodes connected via pointers. Efficient insertion/deletion at known positions but O(n) access.

### Key Concepts
- Node structure (val + next pointer)
- Singly vs doubly linked
- Dummy head for edge cases
- Fast/slow pointer technique
- In-place reversal

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Reverse Linked List | Easy | Three pointers |
| Detect Cycle | Easy | Fast/slow |
| Merge Two Sorted | Easy | Dummy head |
| Remove Nth From End | Medium | Two pointers |
| Reorder List | Medium | Find mid + reverse |

### Real-World Applications
- Undo/redo functionality
- Browser history
- LRU cache implementation
- Memory allocation (free lists)

## Stacks & Queues

**Directory:** `stacks_queues/`

Abstract data types with restricted access. Stacks are LIFO (Last In First Out), queues are FIFO (First In First Out).

### Key Concepts
- Stack: push/pop/peek O(1)
- Queue: enqueue/dequeue O(1)
- Monotonic stack for next greater/smaller
- Deque for sliding window max/min
- Two stacks implement queue

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Valid Parentheses | Easy | Stack matching |
| Min Stack | Medium | Two stacks |
| Daily Temperatures | Medium | Monotonic stack |
| Sliding Window Max | Hard | Monotonic deque |
| Evaluate RPN | Medium | Stack evaluation |

### Real-World Applications
- Function call stack
- Expression parsing
- Task scheduling (queues)
- Message queues (RabbitMQ, Kafka)

## Complexity Comparison

| Structure | Access | Search | Insert (end) | Insert (mid) | Delete |
|-----------|--------|--------|--------------|--------------|--------|
| Array | O(1) | O(n) | O(1)* | O(n) | O(n) |
| Linked List | O(n) | O(n) | O(1)** | O(1)*** | O(1)*** |
| Stack | O(n) | O(n) | O(1) | N/A | O(1) |
| Queue | O(n) | O(n) | O(1) | N/A | O(1) |

*Amortized for dynamic arrays 
**With tail pointer 
***If position is known

## Pattern Recognition

| Pattern | Use When | Data Structure |
|---------|----------|----------------|
| Two Pointers | Sorted array, palindrome | Array |
| Sliding Window | Subarray/substring with constraint | Array |
| Prefix Sum | Range sum queries | Array |
| Hash Map | O(1) lookup needed | Array |
| Fast/Slow | Cycle detection, find middle | Linked List |
| Dummy Head | Simplify edge cases | Linked List |
| Monotonic Stack | Next greater/smaller element | Stack |
| Matching Pairs | Balanced brackets, tags | Stack |
| BFS | Level-order traversal | Queue |
