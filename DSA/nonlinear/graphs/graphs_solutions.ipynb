## Q1: Number of Islands

### Approach: DFS/BFS Flood Fill
- Iterate through grid, when we find '1' (land), increment count
- Use DFS to mark all connected land as visited (change to '0')
- Each DFS marks one complete island

### Complexity
- **Time:** O(m × n) - visit each cell once
- **Space:** O(m × n) - recursion stack in worst case

```python
def num_islands(grid: List[List[str]]) -> int:
    if not grid:
        return 0
    
    rows, cols = len(grid), len(grid[0])
    count = 0
    
    def dfs(r, c):
        # Out of bounds or water
        if r < 0 or r >= rows or c < 0 or c >= cols or grid[r][c] == '0':
            return
        
        # Mark as visited
        grid[r][c] = '0'
        
        # Explore all 4 directions
        dfs(r + 1, c)
        dfs(r - 1, c)
        dfs(r, c + 1)
        dfs(r, c - 1)
    
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == '1':
                count += 1
                dfs(r, c)
    
    return count

# Test
grid = [["1","1","0"],["0","1","0"],["0","0","1"]]
print(num_islands(grid))  # 2
```

## Q2: Clone Graph

### Approach: DFS with Hash Map
- Use hash map to track {original_node: cloned_node}
- For each node, create clone if not exists
- Recursively clone all neighbors

### Complexity
- **Time:** O(V + E) - visit each node and edge once
- **Space:** O(V) - hash map and recursion stack

```python
class Node:
    def __init__(self, val=0, neighbors=None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

def clone_graph(node: Optional[Node]) -> Optional[Node]:
    if not node:
        return None
    
    cloned = {}  # original -> clone
    
    def dfs(n):
        if n in cloned:
            return cloned[n]
        
        # Create clone
        copy = Node(n.val)
        cloned[n] = copy
        
        # Clone neighbors
        for neighbor in n.neighbors:
            copy.neighbors.append(dfs(neighbor))
        
        return copy
    
    return dfs(node)

print("Solution defined!")
```

## Q3: Course Schedule (Cycle Detection)

### Approach: DFS with 3-State Coloring
- 0 = unvisited, 1 = visiting (in current path), 2 = visited
- If we reach a node that's "visiting", we found a cycle

### Complexity
- **Time:** O(V + E) - visit each node and edge once
- **Space:** O(V + E) - adjacency list and recursion stack

```python
def can_finish(numCourses: int, prerequisites: List[List[int]]) -> bool:
    # Build adjacency list
    graph = defaultdict(list)
    for course, prereq in prerequisites:
        graph[prereq].append(course)
    
    # 0 = unvisited, 1 = visiting, 2 = visited
    state = [0] * numCourses
    
    def has_cycle(node):
        if state[node] == 1:  # Currently visiting - cycle!
            return True
        if state[node] == 2:  # Already processed
            return False
        
        state[node] = 1  # Mark as visiting
        
        for neighbor in graph[node]:
            if has_cycle(neighbor):
                return True
        
        state[node] = 2  # Mark as visited
        return False
    
    # Check all nodes (graph may be disconnected)
    for course in range(numCourses):
        if has_cycle(course):
            return False
    
    return True

# Test
print(can_finish(2, [[1, 0]]))  # True
print(can_finish(2, [[1, 0], [0, 1]]))  # False
```

## Q4: Course Schedule II (Topological Sort)

### Approach: Kahn's Algorithm (BFS)
- Calculate in-degree for each node
- Start with nodes that have in-degree 0
- Process node, reduce in-degree of neighbors
- Add neighbors with in-degree 0 to queue

### Complexity
- **Time:** O(V + E)
- **Space:** O(V + E)

```python
def find_order(numCourses: int, prerequisites: List[List[int]]) -> List[int]:
    # Build graph and in-degree count
    graph = defaultdict(list)
    in_degree = [0] * numCourses
    
    for course, prereq in prerequisites:
        graph[prereq].append(course)
        in_degree[course] += 1
    
    # Start with courses that have no prerequisites
    queue = deque([i for i in range(numCourses) if in_degree[i] == 0])
    result = []
    
    while queue:
        course = queue.popleft()
        result.append(course)
        
        for neighbor in graph[course]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)
    
    # If we processed all courses, return result
    return result if len(result) == numCourses else []

# Test
print(find_order(4, [[1,0],[2,0],[3,1],[3,2]]))  # [0, 1, 2, 3] or [0, 2, 1, 3]
```

## Q5: Pacific Atlantic Water Flow

### Approach: Reverse BFS from Oceans
- Instead of checking each cell can reach both oceans
- Start from ocean edges, find all cells that can reach that ocean
- Return intersection of cells that can reach both

### Complexity
- **Time:** O(m × n)
- **Space:** O(m × n)

```python
def pacific_atlantic(heights: List[List[int]]) -> List[List[int]]:
    if not heights:
        return []
    
    rows, cols = len(heights), len(heights[0])
    pacific = set()
    atlantic = set()
    
    def dfs(r, c, visited, prev_height):
        if (r < 0 or r >= rows or c < 0 or c >= cols or 
            (r, c) in visited or heights[r][c] < prev_height):
            return
        
        visited.add((r, c))
        dfs(r + 1, c, visited, heights[r][c])
        dfs(r - 1, c, visited, heights[r][c])
        dfs(r, c + 1, visited, heights[r][c])
        dfs(r, c - 1, visited, heights[r][c])
    
    # Start DFS from ocean edges
    for c in range(cols):
        dfs(0, c, pacific, 0)           # Top edge (Pacific)
        dfs(rows - 1, c, atlantic, 0)   # Bottom edge (Atlantic)
    
    for r in range(rows):
        dfs(r, 0, pacific, 0)           # Left edge (Pacific)
        dfs(r, cols - 1, atlantic, 0)   # Right edge (Atlantic)
    
    # Return intersection
    return [[r, c] for r, c in pacific & atlantic]

print("Solution defined!")
```

## Q6: Rotting Oranges

### Approach: Multi-source BFS
- Start BFS from all rotten oranges simultaneously
- Each level = 1 minute
- Track fresh oranges count

### Complexity
- **Time:** O(m × n)
- **Space:** O(m × n)

```python
def oranges_rotting(grid: List[List[int]]) -> int:
    rows, cols = len(grid), len(grid[0])
    queue = deque()
    fresh = 0
    
    # Find all rotten oranges and count fresh
    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == 2:
                queue.append((r, c))
            elif grid[r][c] == 1:
                fresh += 1
    
    if fresh == 0:
        return 0
    
    minutes = 0
    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]
    
    while queue:
        minutes += 1
        for _ in range(len(queue)):
            r, c = queue.popleft()
            for dr, dc in directions:
                nr, nc = r + dr, c + dc
                if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 1:
                    grid[nr][nc] = 2
                    fresh -= 1
                    queue.append((nr, nc))
    
    return minutes - 1 if fresh == 0 else -1

# Test
print(oranges_rotting([[2,1,1],[1,1,0],[0,1,1]]))  # 4
```

## Q7: Word Ladder

### Approach: BFS with Preprocessing
- Create pattern map: "h*t" -> ["hot", "hat", ...]
- BFS from beginWord, try all patterns

### Complexity
- **Time:** O(M² × N) where M = word length, N = number of words
- **Space:** O(M² × N)

```python
def ladder_length(beginWord: str, endWord: str, wordList: List[str]) -> int:
    if endWord not in wordList:
        return 0
    
    # Build pattern map
    word_set = set(wordList)
    patterns = defaultdict(list)
    for word in wordList:
        for i in range(len(word)):
            pattern = word[:i] + '*' + word[i+1:]
            patterns[pattern].append(word)
    
    # BFS
    queue = deque([(beginWord, 1)])
    visited = {beginWord}
    
    while queue:
        word, length = queue.popleft()
        
        for i in range(len(word)):
            pattern = word[:i] + '*' + word[i+1:]
            
            for neighbor in patterns[pattern]:
                if neighbor == endWord:
                    return length + 1
                
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append((neighbor, length + 1))
    
    return 0

# Test
print(ladder_length("hit", "cog", ["hot","dot","dog","lot","log","cog"]))  # 5
```

## Q8: Graph Valid Tree

### Approach: Union-Find or DFS
- Tree must have exactly n-1 edges
- Tree must be connected (one component)
- Tree must have no cycles

### Complexity
- **Time:** O(V + E)
- **Space:** O(V)

```python
def valid_tree(n: int, edges: List[List[int]]) -> bool:
    # Tree must have exactly n-1 edges
    if len(edges) != n - 1:
        return False
    
    # Build adjacency list
    graph = defaultdict(list)
    for a, b in edges:
        graph[a].append(b)
        graph[b].append(a)
    
    # Check connectivity using DFS
    visited = set()
    
    def dfs(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor)
    
    dfs(0)
    
    # Must visit all nodes
    return len(visited) == n

# Test
print(valid_tree(5, [[0,1],[0,2],[0,3],[1,4]]))  # True
print(valid_tree(5, [[0,1],[1,2],[2,3],[1,3],[1,4]]))  # False (cycle)
```
