# Trees - Solutions

Complete solutions with explanations and complexity analysis.

```python
from collections import deque
from typing import Optional, List

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(vals):
    if not vals or vals[0] is None:
        return None
    root = TreeNode(vals[0])
    queue = deque([root])
    i = 1
    while queue and i < len(vals):
        node = queue.popleft()
        if i < len(vals) and vals[i] is not None:
            node.left = TreeNode(vals[i])
            queue.append(node.left)
        i += 1
        if i < len(vals) and vals[i] is not None:
            node.right = TreeNode(vals[i])
            queue.append(node.right)
        i += 1
    return root
```

## Q1: Maximum Depth of Binary Tree

### Approach: Recursion (DFS)
- Base case: empty tree has depth 0
- Recursive case: depth = 1 + max(left depth, right depth)

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(h) - recursion stack, where h is height (O(n) worst case, O(log n) balanced)

```python
def max_depth(root: Optional[TreeNode]) -> int:
    if not root:
        return 0
    
    left_depth = max_depth(root.left)
    right_depth = max_depth(root.right)
    
    return 1 + max(left_depth, right_depth)

# Test
assert max_depth(build_tree([3, 9, 20, None, None, 15, 7])) == 3
print("Test passed!")
```

## Q2: Invert Binary Tree

### Approach: Recursion
- Swap left and right children
- Recursively invert left and right subtrees

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(h) - recursion stack

```python
def invert_tree(root: Optional[TreeNode]) -> Optional[TreeNode]:
    if not root:
        return None
    
    # Swap children
    root.left, root.right = root.right, root.left
    
    # Recursively invert subtrees
    invert_tree(root.left)
    invert_tree(root.right)
    
    return root

print("Solution defined!")
```

## Q3: Same Tree

### Approach: Recursion
- Both null → same
- One null → not same
- Different values → not same
- Recursively check left and right subtrees

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(h) - recursion stack

```python
def is_same_tree(p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
    # Both empty
    if not p and not q:
        return True
    
    # One empty or different values
    if not p or not q or p.val != q.val:
        return False
    
    # Check subtrees
    return is_same_tree(p.left, q.left) and is_same_tree(p.right, q.right)

print("Solution defined!")
```

## Q4: Symmetric Tree

### Approach: Recursion with Helper
- Tree is symmetric if left subtree is mirror of right subtree
- Two trees are mirrors if:
  - Root values are equal
  - Left of one equals right of other (recursively)

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(h) - recursion stack

```python
def is_symmetric(root: Optional[TreeNode]) -> bool:
    def is_mirror(t1, t2):
        if not t1 and not t2:
            return True
        if not t1 or not t2:
            return False
        return (t1.val == t2.val and 
                is_mirror(t1.left, t2.right) and 
                is_mirror(t1.right, t2.left))
    
    return is_mirror(root, root)

# Test
assert is_symmetric(build_tree([1, 2, 2, 3, 4, 4, 3])) == True
print("Test passed!")
```

## Q5: Validate Binary Search Tree

### Approach: Pass Valid Range Down
- Each node must be within a valid range (min, max)
- For left child: max = parent's value
- For right child: min = parent's value

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(h) - recursion stack

```python
def is_valid_bst(root: Optional[TreeNode]) -> bool:
    def validate(node, low=float('-inf'), high=float('inf')):
        if not node:
            return True
        
        # Value must be within range
        if node.val <= low or node.val >= high:
            return False
        
        # Check subtrees with updated ranges
        return (validate(node.left, low, node.val) and 
                validate(node.right, node.val, high))
    
    return validate(root)

# Test
assert is_valid_bst(build_tree([2, 1, 3])) == True
assert is_valid_bst(build_tree([5, 1, 4, None, None, 3, 6])) == False
print("Tests passed!")
```

## Q6: Binary Tree Level Order Traversal

### Approach: BFS with Queue
- Use queue to process nodes level by level
- For each level, process all nodes currently in queue
- Add their children to queue for next level

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(w) - where w is max width of tree (up to n/2)

```python
def level_order(root: Optional[TreeNode]) -> List[List[int]]:
    if not root:
        return []
    
    result = []
    queue = deque([root])
    
    while queue:
        level = []
        level_size = len(queue)
        
        for _ in range(level_size):
            node = queue.popleft()
            level.append(node.val)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        
        result.append(level)
    
    return result

# Test
assert level_order(build_tree([3, 9, 20, None, None, 15, 7])) == [[3], [9, 20], [15, 7]]
print("Test passed!")
```

## Q7: Lowest Common Ancestor

### Approach: Recursive Search
- If current node is p or q, return it
- Search left and right subtrees
- If both return non-null, current node is LCA
- Otherwise, return whichever is non-null

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(h) - recursion stack

```python
def lowest_common_ancestor(root: TreeNode, p: TreeNode, q: TreeNode) -> TreeNode:
    # Base case: reached null or found p or q
    if not root or root == p or root == q:
        return root
    
    # Search in subtrees
    left = lowest_common_ancestor(root.left, p, q)
    right = lowest_common_ancestor(root.right, p, q)
    
    # If both found, root is LCA
    if left and right:
        return root
    
    # Return whichever is non-null
    return left or right

print("Solution defined!")
```

## Q8: Binary Tree Right Side View

### Approach: BFS (last node of each level)
- Level order traversal
- Take only the last node of each level

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(w) - max width of tree

```python
def right_side_view(root: Optional[TreeNode]) -> List[int]:
    if not root:
        return []
    
    result = []
    queue = deque([root])
    
    while queue:
        level_size = len(queue)
        
        for i in range(level_size):
            node = queue.popleft()
            
            # Last node of this level
            if i == level_size - 1:
                result.append(node.val)
            
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
    
    return result

# Test
assert right_side_view(build_tree([1, 2, 3, None, 5, None, 4])) == [1, 3, 4]
print("Test passed!")
```

## Q9: Serialize and Deserialize Binary Tree

### Approach: Preorder Traversal
- Serialize: preorder traversal, use "null" for None
- Deserialize: rebuild using same preorder sequence

### Complexity
- **Time:** O(n) for both operations
- **Space:** O(n) - storing all nodes

```python
class Codec:
    def serialize(self, root: Optional[TreeNode]) -> str:
        result = []
        
        def preorder(node):
            if not node:
                result.append("null")
                return
            result.append(str(node.val))
            preorder(node.left)
            preorder(node.right)
        
        preorder(root)
        return ",".join(result)
    
    def deserialize(self, data: str) -> Optional[TreeNode]:
        values = iter(data.split(","))
        
        def build():
            val = next(values)
            if val == "null":
                return None
            node = TreeNode(int(val))
            node.left = build()
            node.right = build()
            return node
        
        return build()

print("Solution defined!")
```

## Q10: Kth Smallest Element in BST

### Approach: Inorder Traversal
- Inorder traversal of BST gives sorted order
- Return kth element encountered

### Complexity
- **Time:** O(k) average, O(n) worst case
- **Space:** O(h) - recursion stack

```python
def kth_smallest(root: Optional[TreeNode], k: int) -> int:
    stack = []
    current = root
    count = 0
    
    while stack or current:
        # Go to leftmost
        while current:
            stack.append(current)
            current = current.left
        
        # Process node
        current = stack.pop()
        count += 1
        
        if count == k:
            return current.val
        
        # Move to right subtree
        current = current.right
    
    return -1

# Test
assert kth_smallest(build_tree([3, 1, 4, None, 2]), 1) == 1
print("Test passed!")
```

## Q11: Kth Largest Element in BST

### Approach: Reverse Inorder Traversal
- Reverse inorder (Right → Root → Left) gives descending order
- Return kth element encountered

### Complexity
- **Time:** O(k) average, O(n) worst case
- **Space:** O(h) - recursion stack

```python
def kth_largest(root: Optional[TreeNode], k: int) -> int:
    stack = []
    current = root
    count = 0
    
    while stack or current:
        # Go to rightmost first (reverse inorder)
        while current:
            stack.append(current)
            current = current.right
        
        # Process node
        current = stack.pop()
        count += 1
        
        if count == k:
            return current.val
        
        # Move to left subtree
        current = current.left
    
    return -1

# Test
assert kth_largest(build_tree([3, 1, 4, None, 2]), 1) == 4  # Largest
assert kth_largest(build_tree([5, 3, 6, 2, 4, None, None, 1]), 2) == 5  # 2nd largest
print("Tests passed!")
```

## Q12: Path Sum

### Approach: Recursion (DFS)
- Base case: null node returns False
- At leaf node: check if remaining sum equals node value
- Recursively check left and right with reduced sum

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(h) - recursion stack

```python
def has_path_sum(root: Optional[TreeNode], targetSum: int) -> bool:
    # Empty tree has no path
    if not root:
        return False
    
    # Check if leaf node with matching sum
    if not root.left and not root.right:
        return root.val == targetSum
    
    # Recursively check subtrees with reduced sum
    remaining = targetSum - root.val
    return (has_path_sum(root.left, remaining) or 
            has_path_sum(root.right, remaining))

# Test
root = build_tree([5, 4, 8, 11, None, 13, 4, 7, 2, None, None, None, 1])
assert has_path_sum(root, 22) == True   # 5->4->11->2
assert has_path_sum(root, 26) == True   # 5->8->13
assert has_path_sum(None, 0) == False
print("Tests passed!")
```

## Q13: Path Sum II - Return the Path

### Approach: DFS with Path Tracking
- Maintain current path as we traverse
- At leaf node, if sum matches, return copy of path
- Backtrack by removing node after exploring

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(h) - recursion stack + path storage

```python
def path_sum_return_path(root: Optional[TreeNode], targetSum: int) -> List[int]:
    def dfs(node, remaining, path):
        if not node:
            return []
        
        path.append(node.val)
        
        # Check if leaf with matching sum
        if not node.left and not node.right:
            if remaining == node.val:
                return path.copy()  # Return copy of valid path
            path.pop()  # Backtrack
            return []
        
        # Try left subtree
        result = dfs(node.left, remaining - node.val, path)
        if result:
            return result
        
        # Try right subtree
        result = dfs(node.right, remaining - node.val, path)
        if result:
            return result
        
        path.pop()  # Backtrack
        return []
    
    return dfs(root, targetSum, [])

# Test
root = build_tree([5, 4, 8, 11, None, 13, 4, 7, 2, None, None, None, 1])
assert path_sum_return_path(root, 22) == [5, 4, 11, 2]
assert path_sum_return_path(root, 26) == [5, 8, 13]
print("Tests passed!")
```

## Q14: Path Sum III - Return All Paths

### Approach: DFS Backtracking
- Maintain current path and collect all valid paths
- At leaf node, if sum matches, add copy of path to result
- Always backtrack after exploring both subtrees
- Continue searching even after finding a valid path

### Complexity
- **Time:** O(n) - visit each node once
- **Space:** O(n) - storing all paths in worst case

```python
def path_sum_all_paths(root: Optional[TreeNode], targetSum: int) -> List[List[int]]:
    result = []
    
    def dfs(node, remaining, path):
        if not node:
            return
        
        path.append(node.val)
        
        # Check if leaf with matching sum
        if not node.left and not node.right:
            if remaining == node.val:
                result.append(path.copy())  # Add copy to results
        else:
            # Explore both subtrees
            dfs(node.left, remaining - node.val, path)
            dfs(node.right, remaining - node.val, path)
        
        path.pop()  # Backtrack
    
    dfs(root, targetSum, [])
    return result

# Test
root = build_tree([5, 4, 8, 11, None, 13, 4, 7, 2, None, None, 5, 1])
paths = path_sum_all_paths(root, 22)
assert len(paths) == 2
assert [5, 4, 11, 2] in paths
assert [5, 8, 4, 5] in paths
print("Tests passed!")
```
