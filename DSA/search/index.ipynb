## Binary Search

Divide-and-conquer algorithm that halves the search space with each comparison. Achieves O(log n) time on sorted data.

### Key Concepts
- Requires sorted array or monotonic property
- Three templates: exact match, left boundary, right boundary
- Binary search on answer space
- Rotated array handling
- Peak finding

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Binary Search | Easy | Standard template |
| Search Insert Position | Easy | Left boundary |
| First/Last Position | Medium | Two searches |
| Search Rotated Array | Medium | Two-part analysis |
| Find Peak Element | Medium | Compare neighbors |
| Koko Eating Bananas | Medium | Search on answer |
| Search 2D Matrix | Medium | Flatten or 2D search |
| Find Minimum Rotated | Medium | Compare with right |

### Real-World Applications
- Database indexing (B-trees)
- Git bisect (find buggy commit)
- Dictionary/spell checking
- IP routing tables
- Version compatibility checking

## Search Algorithm Comparison

| Algorithm | Time | Space | Requirements |
|-----------|------|-------|-------------|
| Linear Search | O(n) | O(1) | None |
| Binary Search | O(log n) | O(1) | Sorted array |
| Hash Lookup | O(1)* | O(n) | Hash table built |
| Binary Search Tree | O(log n)* | O(n) | Balanced BST |
| Interpolation Search | O(log log n)* | O(1) | Uniformly distributed |

*Average case

## When to Use Binary Search

### Direct Indicators
- "Sorted array" in problem statement
- "Find minimum/maximum that satisfies condition"
- "Search in rotated sorted array"
- O(log n) time requirement

### Hidden Indicators
- Monotonic function (if f(x) then f(x+1))
- Range of possible answers is known
- "Minimum time/speed/capacity to achieve X"
- Optimization problems with binary decision

## Template Selection Guide

| Problem Type | Template | While Condition | Mid Update |
|--------------|----------|-----------------|------------|
| Find exact value | Template 1 | `left <= right` | `mid ± 1` |
| Find left boundary | Template 2 | `left < right` | `right = mid` |
| Find right boundary | Template 3 | `left < right` | `left = mid` (ceiling) |
| Search on answer | Template 2/3 | `left < right` | Depends on condition |

---

## Common Pitfalls

| Pitfall | Problem | Solution |
|---------|---------|----------|
| Integer overflow | `(left + right) / 2` | `left + (right - left) // 2` |
| Infinite loop | `left = mid` with floor division | Use ceiling: `(left + right + 1) // 2` |
| Off-by-one | Wrong boundary update | Test with small examples |
| Wrong condition | `<=` vs `<` | Match template to problem type |

## Study Tips

1. **Master the three templates** - Know when to use each
2. **Practice on paper** - Draw the search space narrowing
3. **Identify the monotonic property** - What makes binary search applicable?
4. **Test edge cases** - Empty array, single element, duplicates
5. **Binary search on answer** - Powerful technique for optimization problems
