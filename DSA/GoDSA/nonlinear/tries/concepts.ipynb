# Tries (Prefix Trees) - Concepts & Use Cases

A **trie** (pronounced "try") is a tree-based data structure for efficient string storage and retrieval. Also called a **prefix tree** because shared prefixes share the same path.

---
## Core Concept

```
Words: ["cat", "car", "card", "care", "dog"]

 root
 / \
 c d
 | |
 a o
 /| |
 t r g*
 * | 
 /|\
 * d e
 * *

* = end of word marker
```

### Key Properties
- Each node represents a character
- Path from root to node = prefix
- Words with common prefixes share nodes
- End-of-word markers indicate complete words

## Time & Space Complexity

Let:
- m = length of word/prefix
- n = number of words
- L = average word length

| Operation | Time | Hash Table |
|-----------|------|------------|
| Insert | O(m) | O(m) |
| Search | O(m) | O(m) |
| Prefix Search | O(m) | O(n × m) |
| Autocomplete | O(m + k) | O(n × m) |

**Space**: O(n × L × alphabet_size) worst case, but shared prefixes reduce this

---
## Trie vs Hash Table

| Feature | Trie | Hash Table |
|---------|------|------------|
| Prefix search | O(m) | O(n × m) |
| Alphabetical order | Natural | Requires sort |
| Autocomplete | Efficient | Inefficient |
| Space | Can be larger | Usually smaller |
| Longest prefix | O(m) | Complex |

**Use Trie when**:
- Need prefix operations
- Need autocomplete
- Need alphabetical traversal
- Many words share prefixes

## Implementation Variants

### 1. Array-based Children (Fixed Alphabet)
```
class TrieNode:
 def __init__(self):
 self.children = [None] * 26 # a-z only
 self.is_end = False
```
- Faster access: O(1) child lookup
- More memory if sparse

### 2. Dictionary-based Children (Flexible)
```
class TrieNode:
 def __init__(self):
 self.children = {} # Any character
 self.is_end = False
```
- Memory efficient for sparse tries
- Supports any character set

```

# Insert "cat"
trie['c'] = {'a': {'t': {'$': True}}}
```

- Clean and readable

## Real-World Use Cases

### 1. Autocomplete / Type-Ahead
- **Search engines**: Google, Bing suggestions
- **IDEs**: Code completion in VS Code, IntelliJ
- **Mobile keyboards**: Predictive text

```
User types: "prog"
Suggestions: ["program", "progress", "programming", "project"]
```

### 2. Spell Checkers
- Dictionary lookup in O(word_length)
- Find similar words (edit distance + trie)
- Used in MS Word, Grammarly

### 3. IP Routing (Longest Prefix Match)
- Routers use tries for IP lookup
- Find longest matching prefix for destination
- Critical for network performance

```
IP: 192.168.1.100
Routes:
 192.168.0.0/16 → Gateway A
 192.168.1.0/24 → Gateway B (longest match!)
```

### 4. Dictionary / Word Games
- Scrabble word validation
- Boggle word search
- Wordle word checking

### 5. T9 Predictive Text
- Old phone keyboards (2-9 keys)
- Map number sequences to words
- "4663" → "home", "good", "gone"

### 6. DNA Sequence Analysis
- Alphabet: {A, C, G, T}
- Pattern matching in genomes
- Suffix tries for substring search

### 7. File Systems
- Path lookup: `/home/user/documents/file.txt`
- Tab completion in shell
- Directory structure is tree-like

### 8. Contact Search
- Phone contact lookup by name prefix
- Fast filtering as user types

## Trie Variants

### Compressed Trie (Radix Tree / Patricia Trie)
- Merge single-child chains into one node
- Saves space for sparse tries

```
Standard Trie: Compressed:
 r r
 | / \
 o om ub
 / \ | |
 m b ulus ber
 | |
 u b
 | |
 l e
 | |
 u r
 |
 s
```

### Suffix Trie / Suffix Tree
- Store all suffixes of a string
- O(1) substring search after O(n) build
- Used in bioinformatics

### Ternary Search Tree
- Each node has 3 children: less, equal, greater
- More space-efficient than standard trie
- Good for spell checking

---
## Key Tips

1. **Use `setdefault`** for compact trie building
2. **'$' or is_end** to mark word endings
3. **DFS from prefix node** for autocomplete
4. **Recursion with index** for wildcard matching
5. **Word Search II** (board + word list) is classic trie + backtracking
6. Remember: Trie is about **prefixes**, not just storage
