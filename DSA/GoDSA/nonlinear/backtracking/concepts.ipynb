# Backtracking - Concepts & Use Cases

**Backtracking** is an algorithmic technique for solving problems by exploring all potential solutions incrementally, abandoning paths that fail to satisfy constraints ("pruning").

---
## Core Concept

Backtracking = **DFS + Pruning**

```
Problem: Find path from S to E

 S
 /|\
 A B C
 /| |\
X Y Z E ← Found!

Exploration order:
1. S → A → X (dead end, backtrack)
2. S → A → Y (dead end, backtrack)
3. S → A (exhausted, backtrack)
4. S → B (dead end, backtrack)
5. S → C → Z (dead end, backtrack)
6. S → C → E (found!)
```

### Key Idea
1. **Choose**: Make a choice
2. **Explore**: Recursively explore with that choice
3. **Unchoose**: Undo the choice (backtrack)

---
## Template Pattern

```
def backtrack(path, choices):
 # Base case: found valid solution
 if is_solution(path):
 result.append(path.copy()) # Save solution
 return
 
 for choice in choices:
 # Pruning: skip invalid choices early
 if not is_valid(choice):
 continue
 
 # Choose
 path.append(choice)
 
 # Explore
 backtrack(path, remaining_choices)
 
 # Unchoose (backtrack)
 path.pop()
```

## Backtracking vs Other Approaches

| Approach | Description | Use When |
|----------|-------------|----------|
| **Brute Force** | Try all possibilities | Small input, no structure |
| **Backtracking** | Prune invalid paths | Constraints allow early rejection |
| **Dynamic Programming** | Overlapping subproblems | Optimal substructure + overlap |
| **Greedy** | Local optimal choice | Greedy choice property |

### When to Use Backtracking
- Finding **all solutions** (not just one)
- **Constraint satisfaction** problems
- **Combinatorial** problems
- When you can **prune** invalid branches early

## Pruning Strategies

### 1. Constraint Violation
Skip choices that immediately violate constraints.
```
if not is_valid(choice):
 continue # Don't explore this path
```

### 2. Bound Checking
Skip if remaining choices can't lead to valid solution.
```
if remaining_elements < needed:
 return # Can't possibly complete
```

### 3. Duplicate Elimination
Skip duplicates to avoid repeated work.
```
if i > start and nums[i] == nums[i-1]:
 continue # Skip duplicate
```

### 4. Symmetry Breaking
Avoid symmetric solutions.
```
# For N-Queens, only explore first row positions 0 to n//2
```

## Key Tips

1. **Identify the pattern**: Permutation, combination, or subset?
2. **Draw the decision tree** to visualize the recursion
3. **Define state clearly**: What does each recursive call represent?
4. **Find pruning opportunities**: What can be skipped early?
5. **Handle duplicates**: Sort and skip same elements at same level
6. **Don't forget to backtrack**: Undo changes after recursive call
7. **Use `path.copy()`** when saving results (not just `path`)

## Alternative Approach: Include/Exclude Pattern (Knapsack-Style)

Instead of iterating through choices at each level, make a **binary decision** for each element:
1. **Include** the element
2. **Exclude** the element (move to next)

This is similar to the 0/1 Knapsack decision pattern.

```
Traditional Loop: Include/Exclude:
 
for choice in choices: include element i?
 choose(choice) / \
 explore() YES NO
 unchoose(choice) | |
 [include i] [skip i]
 | |
 recurse(i) recurse(i+1)
```

### When to Use This Pattern
- Problems where each element can be **used or skipped**
- **Combination sum** with unlimited use of elements
- **Subset sum** problems
- **Partition** problems
- Problems with **target constraints** (like knapsack)

### Comparison: Loop vs Include/Exclude

| Aspect | Loop Pattern | Include/Exclude |
|--------|--------------|-----------------|
| **Structure** | `for choice in choices` | Binary: include or skip |
| **Index movement** | `start` to track progress | `i` increments on exclude |
| **Reuse elements** | Loop from same `i` | Stay at `i` on include |
| **Best for** | Permutations, combinations | Subset sum, knapsack |
| **Decision tree** | Branching factor = choices | Binary tree |

### When to Choose Each

**Use Loop Pattern:**
- Generating permutations (order matters)
- Combinations with position constraints
- When iterating through choices is natural

**Use Include/Exclude:**
- Subset sum problems
- Partition problems
- 0/1 Knapsack variants
- Problems with "take it or leave it" decisions
- When each element has binary choice

### Include/Exclude Template

```
def backtrack(i, state):
 # Base case: reached end or found solution
 if is_done(i, state):
 return base_result
 
 # Pruning: can't proceed
 if is_invalid(state):
 return invalid_result
 
 # Choice 1: INCLUDE element[i]
 include_result = backtrack(
 i if reuse_allowed else i + 1, # Stay or move
 update_state(state, element[i])
 )
 
 # Choice 2: EXCLUDE element[i]
 exclude_result = backtrack(i + 1, state)
 
 # Combine results (depends on problem)
 return combine(include_result, exclude_result)
```

**Key variations:**
- `i` vs `i + 1`: Stay at same index if element reuse is allowed
- Return type: `bool` (exists?), `int` (count/max), `list` (all solutions)
- Combine: `or` (any), `max/min` (optimize), `append` (collect)
