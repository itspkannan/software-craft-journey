## Trees

**Directory:** `trees/`

Hierarchical data structures with parent-child relationships.

### Key Concepts
- Binary Trees & Binary Search Trees (BST)
- Tree traversals: Preorder, Inorder, Postorder, Level Order
- Balanced trees (AVL, Red-Black)
- Tree properties: height, depth, complete, perfect

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Tree Traversals | Easy | DFS/BFS |
| Maximum Depth | Easy | Recursion |
| Validate BST | Medium | Inorder / Range check |
| Lowest Common Ancestor | Medium | Parent tracking |
| Serialize/Deserialize | Hard | Preorder + markers |

### Real-World Applications
- File systems, DOM trees, database indexes (B-Trees)
- Expression parsing, decision trees, org charts

## Graphs

**Directory:** `graphs/`

Networks of vertices connected by edges, with or without direction/weights.

### Key Concepts
- Representations: Adjacency list, adjacency matrix, edge list
- Traversals: BFS (shortest path), DFS (exploration)
- Directed vs undirected, weighted vs unweighted
- Cycle detection, connected components, topological sort

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Number of Islands | Medium | DFS/BFS flood fill |
| Clone Graph | Medium | BFS + hashmap |
| Course Schedule | Medium | Topological sort |
| Shortest Path | Medium | BFS / Dijkstra |
| Word Ladder | Hard | BFS level traversal |

### Real-World Applications
- Social networks, maps/navigation, web crawling
- Network routing, dependency resolution, recommendation systems

## Heaps

**Directory:** `heaps/`

Complete binary trees maintaining heap property (min or max at root).

### Key Concepts
- Min heap vs max heap
- Array representation (parent/child index formulas)
- Priority queue abstraction
- `heapq` module (min heap only, negate for max)

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Kth Largest Element | Medium | Min heap of size K |
| Top K Frequent | Medium | Counter + heap |
| Merge K Sorted Lists | Hard | Min heap of heads |
| Find Median Stream | Hard | Two heaps (lo/hi) |
| Task Scheduler | Medium | Greedy + heap |

### Real-World Applications
- Task scheduling, Dijkstra's algorithm, event simulation
- Load balancing, streaming median, Huffman coding

## Tries

**Directory:** `tries/`

Tree-like structures for efficient string storage and prefix operations.

### Key Concepts
- Prefix tree structure (shared prefixes = shared nodes)
- TrieNode with children dict and end marker
- Compact dictionary-based implementation
- Variants: compressed trie, suffix trie

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Implement Trie | Medium | Dict of children |
| Word Search II | Hard | Trie + backtracking |
| Add & Search Words | Medium | DFS for wildcards |
| Replace Words | Medium | Prefix lookup |
| Autocomplete | Medium | DFS from prefix |

### Real-World Applications
- Autocomplete, spell checkers, IP routing
- T9 predictive text, DNA analysis, contact search

## Backtracking

**Directory:** `backtracking/`

Algorithmic technique exploring all solutions via incremental construction with pruning.

### Key Concepts
- Choose → Explore → Unchoose pattern
- Decision tree visualization
- Pruning strategies (constraint violation, bounds, duplicates)
- State management and restoration

### Topics Covered
| Problem | Difficulty | Key Technique |
|---------|------------|---------------|
| Subsets | Medium | Include/exclude each |
| Permutations | Medium | Used array |
| Combinations | Medium | Start index |
| N-Queens | Hard | Column/diagonal sets |
| Sudoku Solver | Hard | Cell-by-cell validation |

### Real-World Applications
- Puzzle solvers, game AI (chess, sudoku)
- Constraint satisfaction, scheduling, compiler parsing

## Complexity Comparison

| Structure | Access | Search | Insert | Delete | Space |
|-----------|--------|--------|--------|--------|-------|
| BST (balanced) | O(log n) | O(log n) | O(log n) | O(log n) | O(n) |
| Heap | O(1)* | O(n) | O(log n) | O(log n) | O(n) |
| Trie | - | O(m) | O(m) | O(m) | O(n×m) |
| Graph (adj list) | - | O(V+E) | O(1) | O(E) | O(V+E) |

*Heap access is O(1) for min/max only

## Tips

### When to Use Each Structure

| Pattern | Data Structure |
|---------|----------------|
| Hierarchical data | Tree |
| Sorted data with fast ops | BST |
| K largest/smallest | Heap |
| Prefix matching | Trie |
| Relationships/connections | Graph |
| All possible solutions | Backtracking |
| Shortest path (unweighted) | Graph BFS |
| Shortest path (weighted) | Dijkstra (Heap) |

### Common Patterns
- **Tree/Graph**: DFS recursion, BFS with queue, visited set
- **Heap**: Maintain size K, two heaps for median
- **Trie**: Build from words, DFS for completions
- **Backtracking**: Template with choose/explore/unchoose
