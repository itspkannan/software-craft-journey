# Graphs - Concepts & Use Cases

A **graph** is a data structure consisting of vertices (nodes) and edges (connections between nodes). Unlike trees, graphs can have cycles and multiple paths between nodes.

---
## Core Terminology

```
 A --- B
 |\ |
 | \ |
 | \ |
 D \ C
 \
 E
```

| Term | Definition |
|------|------------|
| **Vertex/Node** | Entity in the graph (A, B, C, D, E) |
| **Edge** | Connection between vertices |
| **Directed** | Edges have direction (A → B) |
| **Undirected** | Edges are bidirectional (A — B) |
| **Weighted** | Edges have associated costs/distances |
| **Degree** | Number of edges connected to a vertex |
| **Path** | Sequence of vertices connected by edges |
| **Cycle** | Path that starts and ends at same vertex |
| **Connected** | Path exists between every pair of vertices |
| **DAG** | Directed Acyclic Graph (no cycles) |

---
## Graph Types

### Undirected Graph
```
 1 --- 2
 | |
 4 --- 3
```
Edges have no direction. If A connects to B, then B connects to A.

### Directed Graph (Digraph)
```
 1 --> 2
 ↓ ↓
 4 <-- 3
```
Edges have direction. A → B doesn't mean B → A.

### Weighted Graph
```
 1 --5-- 2
 | |
 2 3
 | |
 4 --1-- 3
```
Edges have associated weights (costs, distances).

### Special Graphs
- **Tree**: Connected acyclic graph
- **DAG**: Directed graph with no cycles (dependency ordering)
- **Bipartite**: Vertices can be split into two sets with edges only between sets
- **Complete Graph**: Edge between every pair of vertices

## Graph Representations

### 1. Adjacency List (Most Common)
```
graph = {
 'A': ['B', 'C'],
 'B': ['A', 'D'],
 'C': ['A', 'D'],
 'D': ['B', 'C']
}
```
- **Space**: O(V + E)
- **Best for**: Sparse graphs, traversals

### 2. Adjacency Matrix
```
 A B C D
A [ 0 1 1 0 ]
B [ 1 0 0 1 ]
C [ 1 0 0 1 ]
D [ 0 1 1 0 ]
```
- **Space**: O(V²)
- **Best for**: Dense graphs, quick edge lookup

### 3. Edge List
```
edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D')]
```
- **Space**: O(E)
- **Best for**: Algorithms that iterate over all edges (Kruskal's)

## Graph Traversals

### BFS (Breadth-First Search)
- Explores level by level
- Uses **queue**
- Finds **shortest path** in unweighted graphs

```
Start: A
Order: A → B → C → D → E

 A ←─ Start
 / \
 B C ←─ Level 1
 | |
 D E ←─ Level 2
```

### DFS (Depth-First Search)
- Explores as deep as possible first
- Uses **stack** (or recursion)
- Good for pathfinding, cycle detection

```
Start: A
Order: A → B → D → C → E

 A ←─ Start
 / \
 B C
 | |
 D E
 ↓
 Go deep first!
```

## Key Algorithms

### Shortest Path

| Algorithm | Graph Type | Time | Use Case |
|-----------|------------|------|----------|
| **BFS** | Unweighted | O(V+E) | Shortest path in unweighted |
| **Dijkstra** | Non-negative weights | O((V+E) log V) | GPS navigation |
| **Bellman-Ford** | Any weights | O(VE) | Handles negative weights |
| **Floyd-Warshall** | All pairs | O(V³) | All pairs shortest path |

### Topological Sort
Linear ordering of vertices in DAG where u comes before v if edge u → v exists.

```
 A → B → D
 ↓ ↓
 C → E

Topological orders: A, B, C, D, E or A, C, B, E, D
```

### Cycle Detection
- **Undirected**: DFS with parent tracking
- **Directed**: DFS with recursion stack (white/gray/black)

### Connected Components
- **Undirected**: DFS/BFS from unvisited nodes
- **Directed**: Kosaraju's or Tarjan's algorithm for SCCs

## Common Patterns

### 1. BFS Template (Shortest Path)
```
def bfs_shortest(graph, start, target):
 queue = deque([(start, 0)]) # (node, distance)
 visited = {start}
 
 while queue:
 node, dist = queue.popleft()
 if node == target:
 return dist
 
 for neighbor in graph[node]:
 if neighbor not in visited:
 visited.add(neighbor)
 queue.append((neighbor, dist + 1))
 
 return -1 # Not found
```

### 2. DFS Template (Path Finding / Cycle Detection)
```
def dfs_path(graph, node, target, visited):
 if node == target:
 return True
 
 visited.add(node)
 
 for neighbor in graph[node]:
 if neighbor not in visited:
 if dfs_path(graph, neighbor, target, visited):
 return True
 
 return False
```

## Key Tips

1. **Clarify the graph type**: Directed? Weighted? Can have cycles?
2. **Choose representation**: Usually adjacency list is best
3. **BFS vs DFS**:
 - BFS for shortest path (unweighted)
 - DFS for exploring all paths, backtracking
4. **Don't forget visited set** - prevents infinite loops
5. **Grid problems are graph problems** - treat cells as nodes
6. **Topological sort** whenever you see dependencies
