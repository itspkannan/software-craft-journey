## Q1: Two Sum

**Problem:** Given an array of integers and a target, return indices of two numbers that add up to target.

### Approach: Hash Map
- Iterate through array, for each number check if `target - num` exists in hash map
- If yes, return indices; if no, store current number and index

### Complexity
- **Time:** O(n) - single pass through array
- **Space:** O(n) - hash map stores up to n elements

```go
import (
	"fmt"
	"reflect"
	"strings"
	"unicode"
)

func mapHasIntInt(m map[int]int, k int) bool {
	_, ok := m[k]
	return ok
}

func mapHasByteInt(m map[byte]int, k byte) bool {
	_, ok := m[k]
	return ok
}

func eq(msg string, a, b interface{}) {
	if !reflect.DeepEqual(a, b) {
		panic(fmt.Sprintf("%s: %#v != %#v", msg, a, b))
	}
}

func isNil(msg string, v interface{}) {
	if v == nil {
		return
	}
	rv := reflect.ValueOf(v)
	switch rv.Kind() {
	case reflect.Ptr, reflect.Map, reflect.Slice, reflect.Interface, reflect.Chan, reflect.Func:
		if rv.IsNil() {
			return
		}
	}
	panic(msg)
}
```

```go
/*
 * --- Reference solution (notation) ---
 * def two_sum(nums: list[int], target: int) -> list[int]:
 *  seen = {} # value -> index
 *  
 *  for i, num in enumerate(nums):
 *  complement = target - num
 *  if complement in seen:
 *  return [seen[complement], i]
 *  seen[num] = i
 *  
 *  return []
 * 
 * # Test Cases
 * assert two_sum([2, 7, 11, 15], 9) == [0, 1]
 * assert two_sum([3, 2, 4], 6) == [1, 2]
 * assert two_sum([3, 3], 6) == [0, 1]
 * print("All tests passed!")
 */

func TwoSum(nums []int, target int) []int {
	seen := make(map[int]int)
	for i := 0; i < len(nums); i++ {
		num := nums[i]
		complement := (target - num)
		if (mapHasIntInt(seen, complement)) {
			return []int{seen[complement], i}
		}
		seen[num] = i
	}
	return []int{}
}

eq("assert failed", TwoSum([]int{2, 7, 11, 15}, 9), []int{0, 1})
eq("assert failed", TwoSum([]int{3, 2, 4}, 6), []int{1, 2})
eq("assert failed", TwoSum([]int{3, 3}, 6), []int{0, 1})
fmt.Println("All tests passed!")
```

## Q3: Best Time to Buy and Sell Stock

**Problem:** Find maximum profit from one buy-sell transaction.

### Approach: Track Minimum Price
- Track the minimum price seen so far
- At each day, calculate potential profit if we sell today
- Track maximum profit

### Complexity
- **Time:** O(n) - single pass
- **Space:** O(1) - only two variables

```go
/*
 * --- Reference solution (notation) ---
 * def max_profit(prices: list[int]) -> int:
 *  min_price = float('inf')
 *  max_profit = 0
 *  
 *  for price in prices:
 *  min_price = min(min_price, price)
 *  profit = price - min_price
 *  max_profit = max(max_profit, profit)
 *  
 *  return max_profit
 * 
 * # Test Cases
 * assert max_profit([7, 1, 5, 3, 6, 4]) == 5
 * assert max_profit([7, 6, 4, 3, 1]) == 0
 * assert max_profit([1, 2]) == 1
 * print("All tests passed!")
 */

func MaxProfit(prices []int) int {
	min_price := 1e308
	max_profit := 0
	for _, price := range prices {
		min_price = min(min_price, price)
		profit := (price - min_price)
		max_profit = max(max_profit, profit)
	}
	return max_profit
}

eq("assert failed", MaxProfit([]int{7, 1, 5, 3, 6, 4}), 5)
eq("assert failed", MaxProfit([]int{7, 6, 4, 3, 1}), 0)
eq("assert failed", MaxProfit([]int{1, 2}), 1)
fmt.Println("All tests passed!")
```

## Q7: Longest Substring Without Repeating Characters

**Problem:** Find length of longest substring without repeating characters.

### Approach: Sliding Window with Hash Map
- Use hash map to store last index of each character
- When duplicate found, move left pointer past the previous occurrence
- Track maximum window size

### Complexity
- **Time:** O(n) - each character visited at most twice
- **Space:** O(min(n, m)) - where m is character set size

```go
/*
 * --- Reference solution (notation) ---
 * def length_of_longest_substring(s: str) -> int:
 *  char_index = {} # character -> last seen index
 *  left = 0
 *  max_len = 0
 *  
 *  for right, char in enumerate(s):
 *  # If char seen and within current window, shrink window
 *  if char in char_index and char_index[char] >= left:
 *  left = char_index[char] + 1
 *  
 *  char_index[char] = right
 *  max_len = max(max_len, right - left + 1)
 *  
 *  return max_len
 * 
 * # Test Cases
 * assert length_of_longest_substring("abcabcbb") == 3
 * assert length_of_longest_substring("bbbbb") == 1
 * assert length_of_longest_substring("pwwkew") == 3
 * print("All tests passed!")
 */

func LengthOfLongestSubstring(s string) int {
	char_index := make(map[byte]int)
	left := 0
	max_len := 0
	for right := 0; right < len(s); right++ {
		char := s[right]
		if ((mapHasByteInt(char_index, char)) && ((char_index[char] >= left))) {
			left = (char_index[char] + 1)
		}
		char_index[char] = right
		max_len = max(max_len, ((right - left) + 1))
	}
	return max_len
}

eq("assert failed", LengthOfLongestSubstring("abcabcbb"), 3)
eq("assert failed", LengthOfLongestSubstring("bbbbb"), 1)
eq("assert failed", LengthOfLongestSubstring("pwwkew"), 3)
fmt.Println("All tests passed!")
```

## Q10: Container With Most Water

**Problem:** Find two lines that form container with maximum water.

### Approach: Two Pointers
- Start with widest container (left=0, right=n-1)
- Area = min(height[left], height[right]) * width
- Move the shorter line inward (only way to potentially increase area)

### Complexity
- **Time:** O(n) - single pass with two pointers
- **Space:** O(1) - only pointers

```go
/*
 * --- Reference solution (notation) ---
 * def max_area(height: list[int]) -> int:
 *  left, right = 0, len(height) - 1
 *  max_area = 0
 *  
 *  while left < right:
 *  # Calculate area
 *  width = right - left
 *  h = min(height[left], height[right])
 *  area = width * h
 *  max_area = max(max_area, area)
 *  
 *  # Move shorter line
 *  if height[left] < height[right]:
 *  left += 1
 *  else:
 *  right -= 1
 *  
 *  return max_area
 * 
 * # Test Cases
 * assert max_area([1, 8, 6, 2, 5, 4, 8, 3, 7]) == 49
 * assert max_area([1, 1]) == 1
 * print("All tests passed!")
 */

func MaxArea(height []int) int {
	left, right := 0, (len(height) - 1)
	max_area := 0
	for ((left < right)) {
		width := (right - left)
		h := min(height[left], height[right])
		area := (width * h)
		max_area = max(max_area, area)
		if ((height[left] < height[right])) {
			left += 1
		} else {
			right -= 1
		}
	}
	return max_area
}

eq("assert failed", MaxArea([]int{1, 8, 6, 2, 5, 4, 8, 3, 7}), 49)
eq("assert failed", MaxArea([]int{1, 1}), 1)
fmt.Println("All tests passed!")
```
