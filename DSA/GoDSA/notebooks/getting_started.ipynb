```go
import "fmt"

type Point struct{ X, Y int }

p := Point{3, 4}
fmt.Println("Point:", p)
```

```go
// Two pointers style on a slice
nums := []int{1, 2, 3, 4, 5}
i, j := 0, len(nums)-1
for i < j {
	fmt.Println(nums[i], nums[j])
	i++
	j--
}
```
